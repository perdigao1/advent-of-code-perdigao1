```python
#filename = "data_test.txt"
filename = "data.txt"
```

```python
with open(filename,"r") as f:
    lines = f.read().splitlines()
```

```python
lines
```

```python
#first line is the initial sequence

sequence0 = lines[0]
sequence0
```

```python
#Lines from index 2 are the 'dictionary'
seq_repl_dict = {}
for i in range(2, len(lines)):
    li = lines[i]
    seq, repl = li.split(' -> ')
    seq_repl_dict[seq]= repl
```

```python
seq_repl_dict
```

```python
#To do replacement I will proceed as follows
#Create a copy of the current sequence and insert spaces in between all the characters
#Then go to each pair, and check if needs replace.
# At the end, remove spaces to get resulting sequence

seq = sequence0
seq
```

```python
char_freq_dict = {}
```

```python
#initialise frequency dictionary
for i in range(len(seq)):
    char = seq[i]
    if not char in char_freq_dict:
        char_freq_dict[char]=1
    else:
        char_freq_dict[char]+=1
```

```python
#Copy sequence and insert spaces between characters
s0 =seq[0]
for i in range(1, len(seq)):
    char += ' '+ seq[i]
print(s0)
```

```python
#Replace as needed
for i in range(len(seq)):
    i2 = 2*i+1

    s2 = seq[i:i+2]
    
    if s2 in seq_repl_dict:
        #do the substitution
        newchar = seq_repl_dict[s2]
        s0 = s0[:i2] + newchar + s0[i2+1:]

        #Also adjust the character count
        if not newchar in char_freq_dict:
            char_freq_dict[newchar]=1
        else:
            char_freq_dict[newchar]+=1
```

```python
steps=10 #Part1
#steps = 40 # part2 Very slow!!
```

```python
seq = sequence0
char_freq_dict = {}

#initialise frequency dictionary
for i in range(len(seq)):
    char = seq[i]
    if not char in char_freq_dict:
        char_freq_dict[char]=1
    else:
        char_freq_dict[char]+=1


for step in range(steps):
    print(f"step {step}")
    s0 =seq[0]
    for i in range(1, len(seq)):
        s0 += ' '+ seq[i]
    #print(s0)

    #Replace as needed
    for i in range(len(seq)):
        i2 = 2*i+1

        s2 = seq[i:i+2]
        
        if s2 in seq_repl_dict:
            #do the substitution
            newchar = seq_repl_dict[s2]
            s0 = s0[:i2] + newchar + s0[i2+1:]

            #Also adjust the character count
            if not newchar in char_freq_dict:
                char_freq_dict[newchar]=1
            else:
                char_freq_dict[newchar]+=1
    
    s0 = s0.strip()

    seq=s0 #prepare for next iteration
    #print(f"current sequence = {seq}")
print(f"final sequence: {seq}")
print(f"char frequency: {char_freq_dict}")
```
