```python
import numpy as np
```

```python
with open("data.txt","r") as f:
    text_lines = f.readlines()
```

```python
text_lines
```

```python
#First line is the drawn balls
balls = list( map(int,text_lines[0].split(',')) )
```

```python
balls
```

```python
#Get the boards one by one

nboard=0
boards=[]
#boardline=0

import re

i=2
while i < len(text_lines):
    
    #print(text_lines[i].strip())
    line = text_lines[i].strip()

    if line=='':
        print("line is empty")
    else:
        #Read 5 lines
        def splitLineToListInts(s):
            s= s.strip()
            #print(s)
            ret  = list(map(int,re.split(' +', s ) ) )
            print(ret)
            return ret

        board=[]
        board.append( splitLineToListInts(text_lines[i] ) )
        i+=1
        board.append( splitLineToListInts(text_lines[i] ) )
        i+=1
        board.append( splitLineToListInts(text_lines[i] ) )
        i+=1
        board.append( splitLineToListInts(text_lines[i] ) )
        i+=1
        board.append( splitLineToListInts(text_lines[i] ) )

        boards.append( board )
        i+=1
        #print(f"i= {i}")
        #print(f"nboards = {len(boards)}")
    i+=1
```

```python
print(boards)
```

```python
def convToOrder(i):
    index = balls.index(i)#
    return index
```

```python
convToOrder(22)
```

```python
boards_np = np.array(boards)
```

```python
boards_np
```

```python
convToOrder_vect = np.vectorize( convToOrder)
```

```python
boards_order = convToOrder_vect(boards_np)
```

```python
boards_order
```

```python
#For each board get the maximum along rows, and max along columns

b_vmin = []

for b in boards_order:
    print(b)
    b_min_a0 = np.amax( b, axis=0 )
    b_min_a1 = np.amax( b, axis=1 )
    b_vmin.append( np.min( [ b_min_a0, b_min_a1]) )
```

```python
b_vmin
```

```python
#Get the loosing board
board_min = np.argmax(b_vmin)
board_min
```

```python
v_order_min  = b_vmin[board_min]
v_order_min
```

```python
board_win = boards[board_min]
```

```python
board_win
```

```python
#With this board then get the score
#Mark those numbers in the winning board that were called
```

```python
called_numbers = np.array(balls)[:v_order_min+1]
called_numbers
```

```python
lastvalue= balls[v_order_min]
lastvalue
```

```python
def hasBeenCalled(v):
    print(f"v={v}")
    index = np.where(called_numbers == v)
    index = index[0]
    print(f"index={index}")
    ret=0
    if index.size>0:
        ret=1
    return ret

hasBeenCalled_vector = np.vectorize(hasBeenCalled)
```

```python
boardwin_hasBeenCalled = hasBeenCalled_vector(board_win)
```

```python
boardwin_hasBeenCalled
```

```python
boardwin_notBeenCalled = -boardwin_hasBeenCalled +1 #inverts
```

```python
boardwin_notBeenCalled
```

```python
sum_notcalled =  np.sum(boardwin_notBeenCalled* board_win)
```

```python
sum_notcalled
```

```python
score = sum_notcalled * lastvalue
score
```
