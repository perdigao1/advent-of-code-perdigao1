```python
def doExplosion(squid_asstr):
    #print(f"squid input : {squid_asstr}")
    mysquid = squid_asstr
    newsquid= mysquid
    hasExploded=False

    #Explosion
    bracklevel = 0
    ipairstart=None
    ipairend = None
    for i in range(len(mysquid)):
        char = mysquid[i]
        if char=="[":
            bracklevel +=1
            ipairstart = i
        elif char == "]":
            if bracklevel>=5:
                ipairend = i

                #It's the first pair to be exploded
                break #leaves the for loop
            else:
                bracklevel-=1
    
    #print(f"ipairstart={ipairstart} , ipairend={ipairend}")
    
    if not ipairend is None:
        #assert not ipairend is None, "Not a valid exploding pair"

        explpair = eval(mysquid[ipairstart:ipairend+1])
        #print(f"explpair : {explpair}")

        leftpart = mysquid[:ipairstart]
        rightpart = mysquid[ipairend+1:]

        #print(f"leftpart : {leftpart}  ;  rightpart : {rightpart}")

        #Check left
        #From the end, locate the first number that it can find
        intEnd = None
        intStart=None
        for i in range(len(leftpart)-1,0,-1):
            char = leftpart[i]
            #print(f"char {i} is {char}")
            try:
                v0 = int(char)
                #print("is integer")
                if intEnd is None:
                    intEnd = i+1
            except:
                #print("not an integer")
                if not intEnd is None:
                    intStart=i+1
                    break
        
        leftpart_after_expl=leftpart
        if not intEnd is None:
            assert not intStart is None , "no valid numbers when checking left"

            leftpart_left = leftpart[:intStart]
            leftpart_rigth = leftpart[intEnd:]
            leftvalue = int(leftpart[intStart:intEnd])
            incleftvalue = explpair[0] + leftvalue
            leftpart_after_expl = leftpart_left+str(incleftvalue) + leftpart_rigth
        
        #print(f"leftpart_after_expl: {leftpart_after_expl}")

        #Right part
        #Going right towards end, locate the first number that it can find
        intStart = None
        intEnd=None
        for i in range(len(rightpart)):
            char = rightpart[i]
            #print(f"char {i} is {char}")
            try:
                v0 = int(char)
                #print("is integer")
                if intStart is None:
                    intStart=i
            except:
                #print("not an integer")
                if not intStart is None:
                    intEnd=i
                    break
        
        rightpart_after_expl = rightpart
        if not intStart is None:
            rightvalue = int(rightpart[intStart:intEnd])
            rightpart_left = rightpart[:intStart]
            rightpart_right = rightpart[intEnd:]
            incrightvalue = rightvalue + explpair[1]
            rightpart_after_expl = rightpart_left+str(incrightvalue) + rightpart_right

        #print(f"rightpart_after_expl: {rightpart_after_expl}")
        
        newsquid = leftpart_after_expl + "0" + rightpart_after_expl

        hasExploded=True
        #print(f"newsquid: {newsquid}")
    
    return newsquid, hasExploded
```

```python
def addSquids(squid0, squid1):
    squid0_list = eval(squid0)
    squid1_list = eval(squid1)
    newsquid = [squid0_list,squid1_list]
    newsquid_str = str(newsquid)
    return newsquid_str
```

```python
#Create splitReplDict
import math
splitReplDict={}
for i in range(10,200):
    
    v0 = math.floor(i/2)
    v1 = math.ceil(i/2)
    pair_s = str([v0,v1])
    splitReplDict[str(i)] = pair_s
    
def doSplit(squid):
    newsquid=squid
    hasChanged=False
    imin = None
    keymin = None
    #Check for any possible number
    for key in splitReplDict.keys():
        ikey = newsquid.find(key)
        if ikey >=0:
            if imin is None:
                imin = ikey
                keymin=key
            else:
                if ikey<imin:
                    imin=ikey
                    keymin = key
    
    if not imin is None:
        newsquid = newsquid.replace(keymin,splitReplDict[keymin],1)
        hasChanged=True
    
    return newsquid, hasChanged
```

```python
def getMagnitude(snail):
    #snail can be a list or a number
    magn = 0
    
    if type(snail) == int:
        magn = snail
    else:
        assert type(snail)==list , "snail is not a list"
        assert len(snail)==2, "len(snail) is not 2"

        magn = 3* getMagnitude(snail[0]) + 2*getMagnitude(snail[1])
    return magn
```
