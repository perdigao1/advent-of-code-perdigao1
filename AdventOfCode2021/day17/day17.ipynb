```python
#target area

#Test
#targ_ymin, targ_ymax = -10, -1
#targ_xmin, targ_xmax = 20, 30

#puzzle input
targ_ymin, targ_ymax = -157, -103
targ_xmin, targ_xmax = 88, 125
```

```python
#Plan: Iterates increasing values of vx and vy
#Increase vx first and scan vy
#If it misses and overshoots, then check next velocity
```

```python
def checkTrajectoryStartingFrom(x,y, vx,vy):
    traj_x = []
    traj_y = []

    targetHit=False
    overshootx=False
    overshooty=False

    while True:

        #doStep()
        x=x+vx
        y=y+vy

        drag =0
        if vx>0:
            drag=-1
        elif vx<0:
            drag=1
        vx += drag

        vy -= 1 #gravity


        #Check if it landed on target area
        traj_x.append(x)
        traj_y.append(y)

        if x>= targ_xmin and x<=targ_xmax and y>=targ_ymin and y<=targ_ymax:
            #Target hit
            targetHit=True
            break
        if x > targ_xmax:
            overshootx=True
            break
        if y < targ_ymin:
            overshooty=True
            break
    
    return targetHit,overshooty,overshootx, traj_y, traj_x
```

```python
#Determine minimum x velocity
import math
vxmin_approx = (-1.0 + math.sqrt( 1.0 + 8.0*float(targ_xmin) ) / 2.0)
vxmin_approx
```

```python
vxmin = math.ceil(vxmin_approx)
vxmin
```

```python
vxmax = targ_xmax
```

```python
checkTrajectoryStartingFrom(0,0,vxmax+1,0)
```

```python
#Starting velocities

targhit_ymaxs = []
targhit_initialvels=[]

vymax = int(max(abs(targ_ymax),abs(targ_ymin))+1)
print(f"vymax = {vymax}")

for vx in range(vxmin, targ_xmax+1):
    for vy in range(-vymax, vymax):

        print(f"vy,vx = {vy},{vx}")

        targetHit,overshooty,overshootx, traj_y, traj_x = checkTrajectoryStartingFrom(0,0,vx,vy)

        if targetHit:
            print("Target hit")
            #store ymax
            traj_y_np = np.array(traj_y)
            ymax = np.max(traj_y_np)
            targhit_ymaxs.append(ymax)
            print(f"ymax = {ymax} , ")
            targhit_initialvels.append((vy,vx))
        else:
            assert overshooty or overshootx , "ERROR: overshoot?"
```
