```python
import numpy as np
```

```python
#Read data
#filename = "data_test.txt"
filename = "data.txt"
with open(filename,"r") as f:
    text_lines = f.readlines()
```

```python
text_lines
```

```python
def get_movement(s):
    s = s.strip()
    orig_dest_s =  s.split('->')
    x0,y0 = orig_dest_s[0].split(',')
    x0 = int(x0)
    y0 = int(y0)

    x1,y1 = orig_dest_s[1].split(',')
    x1 = int(x1)
    y1 = int(y1)

    return x0, y0 , x1, y1
```

```python
movements = list(map(get_movement,text_lines))
```

```python
movements
```

```python
movements_np = np.array(movements)
```

```python
movements_np
```

```python
max_coord = np.max(movements_np) +1
max_coord
```

```python
def sortasc(x0,x1):
    if x0>x1:
        x2 = x0
        x0 = x1
        x1 = x2
    return x0,x1
```

```python
world = np.zeros((max_coord, max_coord), dtype=np.int32)
for i in range(movements_np.shape[0]):
    #Get origin and end
    movement = movements_np[i]
    print(f"movement = {movement}")
    x0, y0, x1, y1  = movement

    #Horizontal or vertical
    if x0 == x1 or y0==y1:
        isHoriz = True #default
        if x0==x1:
            isHoriz = False
        
        if isHoriz:
            print("Horizontal")
            #add ones to the area
            x0,x1 = sortasc(x0,x1)
            world[y0, x0:x1+1] += np.ones((x1-x0+1), dtype=np.int32)
        else: #vertical
            print("Vertical")
            y0,y1 = sortasc(y0,y1)
            world[y0:y1+1, x0] += np.ones((y1-y0+1), dtype=np.int32)
    else:
        print("Ignore movement as it is diagonal")
    print(world)
```

```python
world
```

```python
overlap = np.where(world >=2 ,1,0)
```

```python
overlap_sum = np.sum(overlap)
overlap_sum #ok
```
