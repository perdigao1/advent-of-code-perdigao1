```python
import numpy as np
```

```python
#filename = "input_test.txt"
filename = "input.txt"
```

```python
with open(filename, 'r') as f:
    lines = f.read().splitlines()
```

```python
lines
```

```python
#Get the shape:
shape = (len(lines), len(lines[0]))
shape
```

```python
board = np.zeros(shape, dtype=np.uint8)
```

```python
for iy,l0 in enumerate(lines):
    for ix,c0 in enumerate(l0):
        item = 0
        if c0 =='v' or c0=='V':
            item = 2
        if c0=='>':
            item = 1
        board[iy,ix]=item
```

```python
board
#1 for east facing
#2 for south facing
```

```python
#Try to move east
#Check spaces available
#Perhaps the best way is to get an array with location of east-facing pieces
#Check available location. shift them left to check overlap
# The (available and piece ) are the ones that will move
# Move the pieces by setting locations marked to available, and set the new locations with the piece
```

```python
#Shift board array left
board_left = np.zeros_like(board)
board_left[:,:-1] = board[:,1:]
board_left[:,-1] = board[:,0] #last col, circular
```

```python
board_left
```

```python
ef_tomove = np.bitwise_and( board_left==0 , board==1)
ef_tomove
```

```python
ef_tomove.any()
```

```python
#Shift marked pieces right
pieces_to_move = board*ef_tomove
pieces_to_move_shifted=np.zeros_like(pieces_to_move)
pieces_to_move_shifted[:,1:] = pieces_to_move[:,:-1]
pieces_to_move_shifted[:,0] = pieces_to_move[:,-1]
```

```python
pieces_to_move_shifted
```

```python
newboard = np.bitwise_not(ef_tomove) * board + pieces_to_move_shifted
```

```python
newboard
```

```python
#Move of south facing pieces is similar
```

```python
board = newboard
board
```

```python
#Shift board array up
board_up = np.zeros_like(board)
board_up[:-1,:] = board[1:,:]
board_up[-1,:] = board[0,:] #last row, circular
```

```python
board_up
```

```python
sf_tomove = np.bitwise_and( board_up==0 , board==2)
sf_tomove
```

```python
#Shift marked pieces south
pieces_to_move = board*sf_tomove
pieces_to_move_shifted=np.zeros_like(pieces_to_move)
pieces_to_move_shifted[1:,:] = pieces_to_move[:-1,:]
pieces_to_move_shifted[0,:] = pieces_to_move[-1,:]
```

```python
pieces_to_move_shifted
```

```python
newboard = np.bitwise_not(sf_tomove) * board + pieces_to_move_shifted
```

```python
newboard
```

```python
# Loop east - south moves iteratively

board0=np.copy(board)
step = 0
while True:

    # *** East facing ***
    #Shift board array left
    board_left = np.zeros_like(board0)
    board_left[:,:-1] = board0[:,1:]
    board_left[:,-1] = board0[:,0] #last col, circular

    ef_tomove = np.bitwise_and( board_left==0 , board0==1)

    #Shift marked pieces right
    pieces_to_move = board0*ef_tomove
    pieces_to_move_shifted=np.zeros_like(pieces_to_move)
    pieces_to_move_shifted[:,1:] = pieces_to_move[:,:-1]
    pieces_to_move_shifted[:,0] = pieces_to_move[:,-1]

    newboard = np.bitwise_not(ef_tomove) * board0 + pieces_to_move_shifted

    # *** South facing ***
    board0=newboard

    #Shift board array up
    board_up = np.zeros_like(board0)
    board_up[:-1,:] = board0[1:,:]
    board_up[-1,:] = board0[0,:] #last row, circular

    sf_tomove = np.bitwise_and( board_up==0 , board0==2)

    pieces_to_move = board0*sf_tomove
    pieces_to_move_shifted=np.zeros_like(pieces_to_move)
    pieces_to_move_shifted[1:,:] = pieces_to_move[:-1,:]
    pieces_to_move_shifted[0,:] = pieces_to_move[-1,:]

    newboard = np.bitwise_not(sf_tomove) * board0 + pieces_to_move_shifted

    step+=1

    print(f"step {step}")
    #print(newboard)

    board0=newboard

    #Check if any movement has occured
    hasMovementOccurred = ef_tomove.any() or sf_tomove.any()
    if not hasMovementOccurred:
        break

print(f"number of steps until stop = {step}")
```
