```python
import numpy as np
```

```python
#text_file = open("input_test.txt", "r")
text_file = open("input.txt", "r")
lines = text_file.readlines()
```

```python
input_data=[]

for line in lines:
    values = line.split(' ')
    input_data.append(values)
```

```python
input_data
```

```python
#Track depth and horizontal movement
depth=0
x = 0
aim=0
```

```python
for mov in input_data:
    movvalue = int(mov[1])
    if 'forward' in mov[0]:
        x += movvalue
        depth += aim*movvalue
    if 'down' in mov[0]:
        #depth+= movvalue
        aim += movvalue
    if 'up' in mov[0]:
        #depth-= movvalue
        aim -=movvalue
```
