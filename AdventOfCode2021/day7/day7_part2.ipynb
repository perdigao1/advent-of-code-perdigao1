```python
import numpy as np
```

```python
#filename = "input_test.txt"
filename="input.txt"
```

```python
with open(filename,"r") as f:
    crab_pos = np.fromfile(filename, sep=",",dtype=np.int32)
```

```python
crab_pos
```

```python
#test a value for xp
xp=5
```

```python
crab_pos_diff = np.abs(crab_pos-xp)
fuels = crab_pos_diff*(crab_pos_diff+1)/2 #https://en.wikipedia.org/wiki/1_%2B_2_%2B_3_%2B_4_%2B_⋯
fuels
```

```python
total_fuel= np.sum(fuels)
```

```python
total_fuel
```

```python
def getTotalFuelForPos(xp):
    crab_pos_diff = np.abs(crab_pos-xp)
    fuels = crab_pos_diff*(crab_pos_diff+1)/2 #https://en.wikipedia.org/wiki/1_%2B_2_%2B_3_%2B_4_%2B_⋯
    return np.sum(fuels)
```

```python
#Vectorize previous function
getTotalFuelForPos_vect = np.vectorize(getTotalFuelForPos)
```

```python
#Get the total fuel for all the positions available
xps = np.arange(np.max(crab_pos),dtype=np.int32)

total_fuels = getTotalFuelForPos_vect(xps)
```
