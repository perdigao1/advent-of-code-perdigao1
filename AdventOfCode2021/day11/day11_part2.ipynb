```python
#filename = "data_test.txt"
filename = "data.txt"
```

```python
with open(filename,"r") as f:
    lines = f.read().splitlines() #better than readlines because it removes the return character
```

```python
lines
```

```python
data0=[]
for l in lines:
    l0=l.strip()
    dataline = [int(char) for char in l0]
    data0.append(dataline)
```

```python
data0
```

```python
import numpy as np
data = np.array(data0, dtype=np.int32)
```

```python
data.shape
```

```python
hasFlashed = np.zeros(data.shape, dtype=np.int32)
hasFlashed
```

```python
def incrementElemAt(y,x):
    if hasFlashed[y,x]==0:
        
        data[y,x] += 1 #Increment
        #print(f"increment {y},{x} to {data[y,x]}")

        #Check if flashed and process it
        if data[y,x] >9:
            hasFlashed[y,x] = 1
            #print(f"Flash element {y},{x}")
            #Increment values around

            data[y,x]=0 #Set to zero here

            if y>0: #row above
                incrementElemAt(y-1,x)
                if x>0:
                    incrementElemAt(y-1, x-1)
                if x<9:
                    incrementElemAt(y-1,x+1)
            if y<9: #row below
                incrementElemAt(y+1,x)
                if x>0:
                    incrementElemAt(y+1, x-1)
                if x<9:
                    incrementElemAt(y+1,x+1)
            
            if x>0: #middle row
                incrementElemAt(y, x-1)
            if x<9:
                incrementElemAt(y,x+1)
```

```python
def doStep():
    #Reset flashes
    #hasFlashed = np.zeros(data.shape, dtype=np.int32)
    hasFlashed.fill(0)

    for iy in range(10):
        for ix in range(10):
            print(f"iy,ix = {iy},{ix}")
            incrementElemAt(iy,ix)
    
    #When complete, get the number of flashes

    stepFlashes = np.sum(hasFlashed)
    print(f"stepFlashes = {stepFlashes}")
    return stepFlashes
```

```python
hasAllFlashed=False
i=0

while not hasAllFlashed:
    i+=1
    stepflashes = doStep()
    if stepflashes == 100:
        hasAllFlashed=True

print(f"all flashed at step {i}")
```
