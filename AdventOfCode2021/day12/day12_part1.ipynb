```python
with open(filename,"r") as f:
    lines = f.read().splitlines() #better than readlines because it removes the return character
```

```python
#Define the graph here
class graph():
    vedge_dict = {}

    # def addVertex(self,verticestring):
    #     if not verticestring in self.vertices:
    #         self.vertices.append(verticestring)

    def addEdge(self, edgestring):
        edge_vertices = edgestring.split('-')

        print(f"addEdge, edge_vertices : {edge_vertices}")
        ev0 = edge_vertices[0]
        ev1 = edge_vertices[1]

        #Check first edge
        if ev0 not in self.vedge_dict:
            self.vedge_dict[ev0]=[ev1]
        else:
            vertxs = self.vedge_dict[ev0]
            if ev1 not in vertxs:
                self.vedge_dict[ev0].append(ev1)

        #second edge
        if ev1 not in self.vedge_dict:
            self.vedge_dict[ev1]=[ev0]
        else:
            vertxs = self.vedge_dict[ev1]
            if ev0 not in vertxs:
                self.vedge_dict[ev1].append(ev0)
```

```python
cavegraph =  graph()
```

```python
for l in lines:
    cavegraph.addEdge(l)
```

```python
print( cavegraph.vedge_dict)
```

```python
routeslist=[]
```

```python
#Find paths from start to end

routeslist=[]
#Set start element

def getRoutesFrom(svertex, cummulativeroute, hasVisitedCumm):

    print(f"getRoutesFrom: {svertex} , cummulativeroute: {cummulativeroute}")

    hasVisited=None
    if svertex=='start':
        cummulativeroute=['start']
        #starting, reset hasVisitedCumm dict
        hasVisitedCumm = cavegraph.vedge_dict.copy()
        for el in hasVisitedCumm:
            hasVisitedCumm[el]=False

    routes = cavegraph.vedge_dict[svertex]
    print(f"routes: {routes}")
    
    hasVisited = hasVisitedCumm.copy()
    hasVisited[svertex]=True

    for r in routes:

        if r=='end':
            cumm0 = cummulativeroute.copy()
            cumm0.append(r)
            routeslist.append(cumm0)
            print(f"Route completed : {cumm0}")
        else:
            if r!='start':
                if (r.islower() and not hasVisited[r] ) or not r.islower():
                    cumm0 = cummulativeroute.copy()
                    cumm0.append(r)
                    getRoutesFrom(r, cumm0, hasVisited)
```
