```python
def checkOverlapCount(scandata0, scandata1):
    #Check how many beacons ovelap
    #If the data was translated correctly, then there should be at least one overlap
    count=0
    for b0 in scandata0:
        for b1 in scandata1:
            if np.array_equal(b0,b1):
                count+=1
    #print(f"overlap count = {count}")
    return count
```

```python
def checkOverlapCountOver12(scandata0, scandata1):
    #Check how many beacons ovelap
    #If the data was translated correctly, then there should be at least one overlap
    
    for b1 in scandata1:
        res = np.where( (scandata0 == b1).all(axis=1) )[0]
        if len(res)>0:
            if res[0]>=12:
                return True
    return False
```

```python
def checkFor12Overlaps(scandata0, scandata1):
    #Parameters scandata0 and scandata1 are the scandata from each, as a numpy array
    #Returns the 3D translation that the 2nd volume needs in order to get 12 beacons to match

    #overlap_scan_data = None

    for beacon0 in scandata0:
        for beacon1 in scandata1:
            #Translate scandata in second volume by vector
            translvector = beacon1-beacon0

            scandata1_transl = scandata1 - translvector #TODO: Check it adds to every row

            #print(f"b0:{beacon0} , b1:{beacon1} ; Translate beacon1 by : {translvector}")
            if checkOverlapCount(scandata0,scandata1_transl)>=12:
                print ("Found >12 overlaps, returning the matched scan data")
                return scandata1_transl, translvector
            else:
                #print("No >12 overlap found")
                pass

```

Unfortunately, there's a second problem: the scanners also don't know their rotation or facing direction. Due to magnetic alignment, each scanner is rotated some integer number of 90-degree turns around all of the x, y, and z axes. That is, one scanner might call a direction positive x, while another scanner might call that direction negative y. Or, two scanners might agree on which direction is positive x, but one scanner might be upside-down from the perspective of the other scanner. In total, each scanner could be in any of 24 different orientations: facing positive or negative x, y, or z, and considering any of four directions "up" from that facing.

```python
rotateZBy90Matrix = [   [  0 , -1 ,  0 ] ,
                        [  1 ,  0 ,  0 ] ,
                        [  0 ,  0 ,  1 ]
                    ]

rotateYBy90Matrix = [   [  0 ,  0 ,  1 ] ,
                        [  0 ,  1 ,  0 ] ,
                        [ -1 ,  0 ,  0 ]
                    ]

rotateXByMinus90Matrix = [  [  1 ,  0 ,  0 ] ,
                            [  0 ,  0 ,  1 ] ,
                            [  0 , -1 ,  0 ]
                        ]
```

```python
def applyTransformToScannerData(scannerdata, transfmatrix):
    scannerdata_rot = np.copy(scannerdata)

    for i in range(len(scannerdata)):
        scannerdata_rot[i] = np.matmul(transfmatrix, scannerdata[i])

    return scannerdata_rot
```
