```python
def selectLinesWithCharAtPos(lines, char,ipos):
    linesres = []
    for line in lines:
        line=line.strip()
        #print(line)
        if char in line[ipos]:
            #print('character found')
            linesres.append(line)
    return linesres
```

```python
def getNumberOfOnesAndZerosAtPos(lines,ipos):
    nones=0
    nzeros=0
    for line in lines:
        if line[ipos]=='1':
            nones+=1
        else:
            nzeros+=1
    
    return{ "1" :nones , "0":nzeros}
```

```python
def getFinalElement(lines, mode='most'):
    nlen = len(lines[0])
    npos=0
    
    #Do until only one element is left

    lines0 = lines #initialize
    
    while npos< nlen:

        #Get number of zeros and ones
        n01_dict = getNumberOfOnesAndZerosAtPos(lines0,npos)

        charsel = '0'
        if mode=='most':
            if n01_dict['1']>=n01_dict['0']:
                charsel='1'
        else:
            #least
            if n01_dict['1']<n01_dict['0']:
                charsel='1'


        lines1 = selectLinesWithCharAtPos(lines0,charsel,npos)

        print(f"npos = {npos}")
        print(lines1)
        if len(lines1) <= 1:
            break
        
        npos+=1
        lines0 = lines1
    
    return lines1[0]

```

```python
getLastMost= getFinalElement(lines,'most')
getLastMost
```

```python
getLastLeast = getFinalElement(lines,'least')
getLastLeast
```

```python
ox_rating = int(getLastMost,2)
co2_rating = int(getLastLeast,2)
print (ox_rating)
print(co2_rating)
```

```python
class processLines:
    nvalues=0
    total_ones=None
    nrange=0

    def newLine(self,line):
        line=line.strip()
        if self.nvalues==0:
            #initialise
            self.nrange = len(line)
            self.total_ones = np.zeros((self.nrange), dtype = np.float32)
        
        #Process the line
        for i in range(self.nrange):
            v = int(line[i])
            self.total_ones[i]+=v
        
        self.nvalues+=1
    
    def getMostCommon(self):
        res = np.rint(self.total_ones / self.nvalues) #check ok
        return res
```

```python
pl = processLines()
```

```python
for line in lines:
    #print(line)
    pl.newLine(line)
```

```python
#Get the result

res= pl.getMostCommon()
print(res)
```

```python
print(pl.nrange)
print(pl.nvalues)
print(pl.total_ones)
```
