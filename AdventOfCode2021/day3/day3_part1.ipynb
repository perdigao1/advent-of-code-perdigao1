```python
import numpy as np
```

```python
#text_file = open("data_test.txt", "r")
text_file = open("data.txt", "r")
```

```python
lines = text_file.readlines()
```

```python
lines
```

```python
class processLines:
    nvalues=0
    total_ones=None
    nrange=0

    def newLine(self,line):
        line=line.strip()
        if self.nvalues==0:
            #initialise
            self.nrange = len(line)
            self.total_ones = np.zeros((self.nrange), dtype = np.float32)
        
        #Process the line
        for i in range(self.nrange):
            v = int(line[i])
            self.total_ones[i]+=v
        
        self.nvalues+=1
    
    def getMostCommon(self):
        res = np.rint(self.total_ones / self.nvalues) #check ok
        return res
```

```python
pl = processLines()
```

```python
for line in lines:
    #print(line)
    pl.newLine(line)
```

```python
#Get the result

res= pl.getMostCommon()
print(res)
```

```python
print(pl.nrange)
print(pl.nvalues)
print(pl.total_ones)
```

```python
#Convert to decimal
decvalue_string=''
for v in res:
    if v==1:
        decvalue_string += '1'
    else:
        decvalue_string += '0'
decvalue_string
```

```python
#Convert to decimal
decvalue = int(decvalue_string,2)
decvalue
```

```python
gamma_rate = decvalue
```

```python
#epsillon_rate
#invert
decvalue_string_inv=''
for v in res:
    if v==0:
        decvalue_string_inv += '1'
    else:
        decvalue_string_inv += '0'
decvalue_string_inv
decvalue_inv = int(decvalue_string_inv,2)
epsillon_rate = decvalue_inv
```

```python
power_consumption = gamma_rate * epsillon_rate
```
