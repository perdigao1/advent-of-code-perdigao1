```python
#filename = "data_test.txt"
filename = "data.txt"
```

```python
with open(filename,"r") as f:
    lines = f.read().splitlines() #better than readlines because it removes the return character
```

```python
lines
```

```python
score = 0

invCharacterScoreDict={ ')':3 , ']':57 , '}':1197 , '>':25137 }

for l in lines:

    openchars=[]

    ichar = 0

    isValid=True

    print(f"line : {l}")
    for ichar in range(len(l)):
        char = l[ichar]
        print(f"char:{char}")
        #is it an opening?
        if char=='(' or char=='[' or char=='{' or char=='<':
            print("opening")
            openchars.append(char)
        else:
            #Not an opening char, check it is a closing char
            lastopenchar = openchars[-1]
            print(f"Not opening, perhaps closing. lastopenchar={lastopenchar}")
            score_incr = 0
            if char==')':
                if lastopenchar =='(':
                    #valid
                    print("Valid closing")
                    openchars.pop()
                else:
                    #not valid
                    score_incr=invCharacterScoreDict[char]
                    score+=score_incr
                    print(f"Invalid. score_incr = {score_incr}")
                    break

            if char==']':
                if lastopenchar =='[':
                    #valid
                    print("Valid closing")
                    openchars.pop()
                else:
                    #not valid
                    score_incr=invCharacterScoreDict[char]
                    score+=score_incr
                    print(f"Invalid. score_incr = {score_incr}")
                    break

            if char=='}':
                if lastopenchar =='{':
                #valid
                    print("Valid closing")
                    openchars.pop()
                else:
                    #not valid
                    score_incr=invCharacterScoreDict[char]
                    score+=score_incr
                    print(f"Invalid. score_incr = {score_incr}")
                    break

            if char=='>':
                if lastopenchar =='<':
                    #valid
                    print("Valid closing")
                    openchars.pop()
                else:
                    #not valid
                    score_incr=invCharacterScoreDict[char]
                    score+=score_incr
                    print(f"Invalid. score_incr = {score_incr}")
                    break
```
