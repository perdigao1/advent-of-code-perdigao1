```python
#filename = "data_test.txt"
filename = "data.txt"
```

```python
with open(filename,"r") as f:
    lines = f.read().splitlines() #better than readlines because it removes the return character
```

```python
lines
```

```python
incscores = []

dictCloseToOpen={ ')':'('  ,  ']':'['  ,  '}':'{' , '>':'<'}
invCharacterScoreDict={ ')':3 , ']':57 , '}':1197 , '>':25137 }
incompleteScoreDict = { '(':1 , '[':2 , '{':3 , '<':4 }

for l in lines:

    openchars=[]

    ichar = 0

    isValid=True

    print(f"line : {l}")
    for ichar in range(len(l)):
        char = l[ichar]
        print(f"char:{char}")
        #is it an opening?
        if char=='(' or char=='[' or char=='{' or char=='<':
            print("opening char")
            openchars.append(char)
        else:
            #Not an opening char, check it is a closing char
            lastopenchar = openchars[-1]
            print(f"Not opening, perhaps closing. lastopenchar={lastopenchar}")
            score_incr = 0
            if dictCloseToOpen[char]==lastopenchar:
                print("Valid")
                openchars.pop()
            else:
                print(f"Invalid. Ignoring line")
                isValid=False
                break

    #By the end of the for loop, openchars should have the remaining characters
    #Use them to get the completion score

    if isValid:
        print(f"Closing chars remaining : {openchars}")
        
        openchars.reverse()

        scoreline=0
        if len(openchars) >0:
            print("Incomplete line") 

            for char in openchars:
                scoreline *= 5
                scoreline += incompleteScoreDict[char]
            print(f"Incomplete score {scoreline}")
            incscores.append(scoreline)
```
