```python
import numpy as np
```

```python
#filename = "data_test.txt"
filename = "data.txt"
```

```python
with open(filename, "r") as f:
    lines = f.readlines()
```

```python
lines
```

```python
#Get the right side
leftside = []
rightside = []
for l in lines:
    left,right = l.strip().split(' | ')
    leftside.append(left)
    rightside.append(right)
```

```python
leftside
```

```python
rightside
```

```python
def splitBySpace(s):
    return s.split(' ')
```

```python
outputdigits = np.array( list(map(splitBySpace,rightside)))
```

```python
outputdigits
```

```python
outputdigits_nchars = np.char.str_len(outputdigits)
outputdigits_nchars
```

```python
out_digit1 = np.where(outputdigits_nchars == 2 ,1,0)
out_digit1
```

```python
out_digit4 = np.where(outputdigits_nchars == 4 ,1,0)
out_digit4
```

```python
out_digit7 = np.where(outputdigits_nchars == 3,1,0)
out_digit7
```

```python
out_digit8 = np.where(outputdigits_nchars == 7 ,1,0)
out_digit8
```

```python
print (f"1,4,7,8 : {np.sum(out_digit1)} , {np.sum(out_digit4)}, {np.sum(out_digit7)}, {np.sum(out_digit8)} " )
```

```python
total1478 = np.sum(out_digit1) + np.sum(out_digit4)+np.sum(out_digit7)+np.sum(out_digit8)
total1478 #OK
```
