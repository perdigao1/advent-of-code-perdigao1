```python
import numpy as np
```

```python
#filename = "data_test.txt"
filename = "data.txt"
```

```python
with open(filename, "r") as f:
    lines = f.readlines()
```

```python
#Get the right side
leftside = []
rightside = []
for l in lines:
    left,right = l.strip().split(' | ')
    leftside.append(left)
    rightside.append(right)
```

```python
rightside
```

```python
def splitBySpace(s):
    return s.split(' ')
```

```python
inputdigits = np.array( list(map(splitBySpace,leftside)))
inputdigits
```

```python
outputdigits = np.array( list(map(splitBySpace,rightside)))
```

```python
'''
Digits determination

nchars can tell the following digits

nchar | digit
2       1
3       7
4       4
5       2,3,5
6       0,6,9
7       8

Identification

digit 3, is the only one that shares the same segments as digit 1
digit 2 and 5, compare with segments in 4. Digit 5 shares +1 segment compared with digit 2
digit 2, compare with segments in 4, shares 2 segments
digit 5, compare with segments in 4, shares 3 segments
digit 9, all segments are shared with digit 4. Compare with segments in digit 4, shares all 4 segments
digit 0, Compare with segments in digit 1, shares 2 segments
digit 6, Compare with segments in digit 1, shares 1 segments

'''
```

```python
#Need to determine the translation line-by-line (per entry)
#First line test
inputline0 = inputdigits[0]
inputline0
```

```python
inp_nchar = np.char.str_len(inputline0)
inp_nchar
```

```python
#Find index for obvious ones
d1_arg = np.argwhere(inp_nchar == 2)[0][0]
d4_arg = np.argwhere(inp_nchar == 4)[0][0]
d7_arg = np.argwhere(inp_nchar == 3)[0][0]
d8_arg = np.argwhere(inp_nchar == 7)[0][0]
```

```python
segm4 = inputline0[d4_arg]
segm4_sorted = sorted(segm4)
def compareWithDigit4(segm):
    n=0
    for c in segm:
        if segm4_sorted.count(c)>0:
            n+=1
    return n

segm1_sorted = sorted(inputline0[d1_arg])
def compareWithDigit1(segm):
    n=0
    for c in segm:
        if segm1_sorted.count(c)>0:
            n+=1
    return n
```

```python
nchars_of_5 = np.argwhere(inp_nchar == 5)
nchars_of_5
```

```python
d3_arg=None
d2_arg=None
d5_arg=None

for a in nchars_of_5:
    arg=a[0]
    segm = inputline0[arg]
    #Compare with segments of d1
    n1 = compareWithDigit1(segm)
    if n1==2:
        #Digit is 3
        d3_arg = arg
    else:
        n5 = compareWithDigit4(segm)
        if n5==2:
            #is 2
            d2_arg=arg
        else:
            d5_arg=arg

print(f"arguments of 2,3,5 : {d2_arg,d3_arg,d5_arg}")
```

```python
d0_arg=None
d6_arg=None
d9_arg=None

for a in nchars_of_6:
    arg=a[0]
    segm = inputline0[arg]
    print(f"segm = {segm}")
    #Compare with segments of d1
    n4 = compareWithDigit4(segm)
    print(f"n4 = {n4}")
    if n4==4:
        print("digit is 9")
        d9_arg=arg
    else:
        n1 = compareWithDigit1(segm)
        print(f"n1 = {n1}")
        if n1==1:
            print("digit is 6")
            d6_arg = arg
        else:
            print("digit is 0")
            d0_arg=arg

print(f"arguments of 0,6,9 : {d0_arg,d6_arg,d9_arg}")
```

```python
def sortCharacters(s):
    return ''.join(sorted(s))
```

```python
#Now that segments have been identified, create a dictionary

segm_to_digit_dict ={
    sortCharacters(inputline0[d0_arg]) : '0' ,
    sortCharacters(inputline0[d1_arg]) : '1' ,
    sortCharacters(inputline0[d2_arg]) : '2' ,
    sortCharacters(inputline0[d3_arg]) : '3' ,
    sortCharacters(inputline0[d4_arg]) : '4' ,
    sortCharacters(inputline0[d5_arg]) : '5' ,
    sortCharacters(inputline0[d6_arg]) : '6' ,
    sortCharacters(inputline0[d7_arg]) : '7' ,
    sortCharacters(inputline0[d8_arg]) : '8' ,
    sortCharacters(inputline0[d9_arg]) : '9'
}
```

```python
def convertSegmToDigits(segm):
    s0 = sortCharacters(segm)
    return(segm_to_digit_dict[s0])
```

# Run from here after getting lines

```python
#Do this for all the lines in the data
values = []

for line in lines:
    print(f"line: {line}")
    left,right = line.strip().split(' | ')
    inputline0 = left.split(' ')
    outputline0 = right.split(' ')

    inp_nchar = np.char.str_len(inputline0)
    d1_arg = np.argwhere(inp_nchar == 2)[0][0]
    d4_arg = np.argwhere(inp_nchar == 4)[0][0]
    d7_arg = np.argwhere(inp_nchar == 3)[0][0]
    d8_arg = np.argwhere(inp_nchar == 7)[0][0]
    print(f"arguments of d1, d4, d7, d8 : {d1_arg},{ d4_arg},{d7_arg},{d8_arg}")

    segm4 = inputline0[d4_arg]
    segm4_sorted = sorted(segm4)
    def compareWithDigit4(segm):
        n=0
        for c in segm:
            if segm4_sorted.count(c)>0:
                n+=1
        return n

    segm1_sorted = sorted(inputline0[d1_arg])
    def compareWithDigit1(segm):
        n=0
        for c in segm:
            if segm1_sorted.count(c)>0:
                n+=1
        return n
    
    nchars_of_5 = np.argwhere(inp_nchar == 5)
    #arg_nchar5_0 = nchars_of_5[0][0]
    #segm5 = inputline0[arg_nchar5_0]

    d3_arg=None
    d2_arg=None
    d5_arg=None

    for a in nchars_of_5:
        arg=a[0]
        segm = inputline0[arg]
        #Compare with segments of d1
        n1 = compareWithDigit1(segm)
        if n1==2:
            #Digit is 3
            d3_arg = arg
        else:
            n5 = compareWithDigit4(segm)
            if n5==2:
                #is 2
                d2_arg=arg
            else:
                d5_arg=arg
    print(f"arguments of 2,3,5 : {d2_arg,d3_arg,d5_arg}")

    nchars_of_6 = np.argwhere(inp_nchar == 6)

    d0_arg=None
    d6_arg=None
    d9_arg=None

    for a in nchars_of_6:
        arg=a[0]
        segm = inputline0[arg]
        print(f"segm = {segm}")
        #Compare with segments of d1
        n4 = compareWithDigit4(segm)
        print(f"n4 = {n4}")
        if n4==4:
            print("digit is 9")
            d9_arg=arg
        else:
            n1 = compareWithDigit1(segm)
            print(f"n1 = {n1}")
            if n1==1:
                print("digit is 6")
                d6_arg = arg
            else:
                print("digit is 0")
                d0_arg=arg

    print(f"arguments of 0,6,9 : {d0_arg,d6_arg,d9_arg}")

    def _sortCharacters(s):
        return ''.join(sorted(s))

    segm_to_digit_dict ={
        _sortCharacters(inputline0[d0_arg]) : '0' ,
        _sortCharacters(inputline0[d1_arg]) : '1' ,
        _sortCharacters(inputline0[d2_arg]) : '2' ,
        _sortCharacters(inputline0[d3_arg]) : '3' ,
        _sortCharacters(inputline0[d4_arg]) : '4' ,
        _sortCharacters(inputline0[d5_arg]) : '5' ,
        _sortCharacters(inputline0[d6_arg]) : '6' ,
        _sortCharacters(inputline0[d7_arg]) : '7' ,
        _sortCharacters(inputline0[d8_arg]) : '8' ,
        _sortCharacters(inputline0[d9_arg]) : '9'
    }

    def _convertSegmToDigits(segm):
        s0 = _sortCharacters(segm)
        return(segm_to_digit_dict[s0])

    digits = list(map(_convertSegmToDigits,outputline0))
    print(f"converted out digits: {digits}")
    value = int(''.join(digits))
    
    values.append(value)
```

```python
#Sum of outvalues
values_np = np.array(values)
np.sum(values_np) #OK
```
