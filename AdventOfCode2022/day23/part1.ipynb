Elves suggest and make moves

Try to distinguish elves, each of them with a list of two numbers (tuple?) representing coordinates.

All elves positions are contained in a list

Start by reading file

```python
#input_filename = './input_test_small.txt'
#input_filename = './input_test.txt'
input_filename='./input.txt'
```

```python
with open(input_filename,'r') as f:
    lines = f.readlines()
```

```python
elves=[]

for i, line in enumerate(lines):
    positions_of_char_in_string = [j for j,char in enumerate(line) if char=='#']

    for pos0 in positions_of_char_in_string:
        elves.append([i,pos0])
    
```

Test running a round

```python
propose_order = ['N','S','W','E']
```

```python
elves_propose_dir=[]
elves_proposed_dest=[]
for e0 in elves:
    print('elf ', e0)
    is_e_N, is_e_NE, is_e_NW =False,False,False
    is_e_S, is_e_SE, is_e_SW = False,False,False
    is_e_E, is_e_W = False,False
    #Check if there are any elves nearby. If no elves, then don't move

    #check N
    ch_N = [ e0[0]-1, e0[1] ]
    if ch_N in elves:
        is_e_N=True
        #print(f"There is a elf N of {e0}")
    is_e_NE= ( [e0[0]-1, e0[1]+1] in elves)
    is_e_NW= ( [e0[0]-1, e0[1]-1] in elves)
    is_e_S= ( [e0[0]+1, e0[1]] in elves)
    is_e_SE= ( [e0[0]+1, e0[1]+1] in elves)
    is_e_SW= ( [e0[0]+1, e0[1]-1] in elves)
    is_e_E= ( [e0[0], e0[1]+1] in elves)
    is_e_W= ( [e0[0], e0[1]-1] in elves)

    prop0=None
    prop_dest0=None
    if not (is_e_N or is_e_NE or is_e_NW or is_e_S or is_e_SE or is_e_SW or is_e_E or is_e_W):
        #No elves around
        prop0=None
    else:
        #Go thorough the order
        for ord0 in propose_order:
            if ord0=='N':
                #Check if there is any elf in the N's directions
                if not (is_e_N or is_e_NE or is_e_NW):
                    prop0='N'
                    prop_dest0=[e0[0]-1,e0[1]]
                    break
            if ord0=='S':
                if not (is_e_S or is_e_SE or is_e_SW):
                    prop0='S'
                    prop_dest0=[e0[0]+1,e0[1]]
                    break
            if ord0=='E':
                if not (is_e_E or is_e_SE or is_e_NE):
                    prop0='E'
                    prop_dest0=[e0[0],e0[1]+1]
                    break
            if ord0=='W':
                if not (is_e_W or is_e_SW or is_e_NW):
                    prop0='W'
                    prop_dest0=[e0[0],e0[1]-1]
                    break
    print('proposed direction:', prop0)
    elves_propose_dir.append('prop0')
    elves_proposed_dest.append(prop_dest0)
```

Second Half. Do the movements. If two elves propose to move to the same tile, none of them moves.

```python
elves1=[]
#epd = elf proposed destination
for i, epd0 in enumerate(elves_proposed_dest):
    # By default, elf stays in same position
    newpos= elves[i]
    
    if not epd0 is None:
        elves_proposed_dest1 = elves_proposed_dest.copy()
        elves_proposed_dest1.pop(i)

        if not epd0 in elves_proposed_dest1:
            #Not a common destination so make the move
            newpos=epd0
    
    elves1.append(newpos)
```

```python
#Helper function to print the location of all elves
import numpy as np
def print_elves(elves):
    elv_np=np.array(elves)
    ymax, xmax=np.max(elv_np, axis=0)
    ymin, xmin=np.min(elv_np, axis=0)

    #print(f"ymax:{ymax}, ymin:{ymin}, xmax:{xmax}, xmin:{xmin}")
    ydim=int(ymax-ymin)+3
    xdim=int(xmax-xmin)+3
    elv_loc= np.zeros((ydim,xdim), dtype=bool)
    
    for e0 in elves:
        elv_loc[e0[0]-ymin+1, e0[1]-xmin+1 ] = True
    
    for irow in range(ydim):
        rowvals = elv_loc[irow,:]

        s=''
        for val in rowvals:
            schar='.'
            if val:
                schar='#'
            s+=schar
        print(s)
```

```python
def get_empty_from_elves(elves):
    elv_np=np.array(elves)
    ymax, xmax=np.max(elv_np, axis=0)
    ymin, xmin=np.min(elv_np, axis=0)

    #print(f"ymax:{ymax}, ymin:{ymin}, xmax:{xmax}, xmin:{xmin}")
    ydim=int(ymax-ymin)+1
    xdim=int(xmax-xmin)+1

    area= ydim*xdim

    return area-len(elves)
```

```python
niterations=10

propose_order_deq = deque(propose_order)

elves0= elves.copy()

print('initial state')
print_elves(elves0)

for it in range(niterations):

    #1st half
    elves_propose_dir=[]
    elves_proposed_dest=[]
    for e0 in elves0:
        #print('elf ', e0)
        is_e_N, is_e_NE, is_e_NW =False,False,False
        is_e_S, is_e_SE, is_e_SW = False,False,False
        is_e_E, is_e_W = False,False
        #Check if there are any elves nearby. If no elves, then don't move

        #check N
        ch_N = [ e0[0]-1, e0[1] ]
        if ch_N in elves0:
            is_e_N=True
            #print(f"There is a elf N of {e0}")
        is_e_NE= ( [e0[0]-1, e0[1]+1] in elves0)
        is_e_NW= ( [e0[0]-1, e0[1]-1] in elves0)
        is_e_S= ( [e0[0]+1, e0[1]] in elves0)
        is_e_SE= ( [e0[0]+1, e0[1]+1] in elves0)
        is_e_SW= ( [e0[0]+1, e0[1]-1] in elves0)
        is_e_E= ( [e0[0], e0[1]+1] in elves0)
        is_e_W= ( [e0[0], e0[1]-1] in elves0)

        prop0=None
        prop_dest0=None
        if not (is_e_N or is_e_NE or is_e_NW or is_e_S or is_e_SE or is_e_SW or is_e_E or is_e_W):
            #No elves around
            prop0=None
        else:
            #Go thorough the order
            for ord0 in propose_order_deq:
                if ord0=='N':
                    #Check if there is any elf in the N's directions
                    if not (is_e_N or is_e_NE or is_e_NW):
                        prop0='N'
                        prop_dest0=[e0[0]-1,e0[1]]
                        break
                if ord0=='S':
                    if not (is_e_S or is_e_SE or is_e_SW):
                        prop0='S'
                        prop_dest0=[e0[0]+1,e0[1]]
                        break
                if ord0=='E':
                    if not (is_e_E or is_e_SE or is_e_NE):
                        prop0='E'
                        prop_dest0=[e0[0],e0[1]+1]
                        break
                if ord0=='W':
                    if not (is_e_W or is_e_SW or is_e_NW):
                        prop0='W'
                        prop_dest0=[e0[0],e0[1]-1]
                        break
        #print('proposed direction:', prop0)
        elves_propose_dir.append('prop0')
        elves_proposed_dest.append(prop_dest0)
    

    #2nd half
    elves1=[]
    #epd = elf proposed destination
    for i, epd0 in enumerate(elves_proposed_dest):
        # By default, elf stays in same position
        newpos= elves0[i]
        
        if not epd0 is None:
            elves_proposed_dest1 = elves_proposed_dest.copy()
            elves_proposed_dest1.pop(i)

            if not epd0 in elves_proposed_dest1:
                #Not a common destination so make the move
                newpos=epd0
        
        elves1.append(newpos)

    #Print result after iteration
    print(f'iteration:{it+1}')
    print_elves(elves1)
    print()
    
    #prepare for next iteration
    elves0=elves1.copy()
    propose_order_deq.rotate(-1)
```
