```python
#file_input="input_test.txt"
file_input="input.txt"
```

```python
with open(file_input,'r') as f:
    lines = f.readlines()
```

```python
scores=[]
for line0 in lines:
    line_split = line0.split(" ")
    #print(line_split)

    choice_score = 0
    if ('X' in line_split[1]):
        choice_score=1
    elif ('Y' in line_split[1]):
        choice_score=2
    elif ('Z' in line_split[1]):
        choice_score=3
    
    result=0 #lost (default)

    #wins
    if ( ('X' in line_split[1]) and ('C' in line_split[0]) ) or ( ('Y' in line_split[1]) and ('A' in line_split[0]) ) or ( ('Z' in line_split[1]) and ('B' in line_split[0]) ):
        result=6 #win
    
    if ( ('X' in line_split[1]) and ('A' in line_split[0]) ) or ( ('Y' in line_split[1]) and ('B' in line_split[0]) ) or ( ('Z' in line_split[1]) and ('C' in line_split[0]) ):
        result=3 #draw

    score = result + choice_score

    print(f"line0:{line0} , choice_score:{choice_score} , result:{result}, score:{score}")
    scores.append(score)
```

```python
sum_scores = np.sum(np.array(scores))
sum_scores
```
