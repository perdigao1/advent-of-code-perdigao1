```python
import numpy as np
```

```python
#file_input="input_test.txt"
file_input="input.txt"
```

```python
with open(file_input,'r') as f:
    lines = f.readlines()
```

ABC
x,y,z
rock,paper, scissors
0,1,2


```python
xyz_to_i = {'X':0 , 'Y':1, 'Z':2}
abc_to_i = {'A':0 , 'B':1, 'C':2}
```

```python
choice_to_scores={0:1 , 1:2, 2:3}
```

```python
scores=[]
for line0 in lines:
    line_split = line0.split(" ")
    #print(line_split)

    col0=0
    col1=0

    #Use dict to convert rock paper scissors to indexes 0,1,2
    col0= abc_to_i[line_split[0]]
    col1= xyz_to_i[line_split[1].strip()]

    choice=col0 #default, need to draw
    result_score = 3
    if col1==0:
        #Need to loose, make a choice that is lower than first
        choice = col0-1
        if choice==-1:
            choice=2
        result_score = 0
    elif col1==2:
        #need to win
        choice=col0+1
        if choice>2:
            choice=0
        result_score = 6

    
    #Now make choice based on 
    choice_score = choice_to_scores[choice]

    score = result_score + choice_score

    print(f"line0:{line0} , choice_score:{choice_score} , result:{result_score}, score:{score}")
    scores.append(score)
```

```python
sum_scores = np.sum(np.array(scores))
sum_scores
```
