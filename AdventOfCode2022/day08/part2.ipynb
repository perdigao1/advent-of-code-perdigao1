```python
import numpy as np
```

```python
#inputfile='input_test.txt'
inputfile='input.txt'

with open(inputfile,'r') as f:
    lines=f.readlines()
lines = [line0.strip() for line0 in lines]
```

```python
lines
```

```python
trees = []
for line0 in lines:
    line_digits = [ int(char) for char in line0 ]
    trees.append(line_digits)
```

```python
trees_np = np.array(trees)
```

```python
def get_dir_tree_score_at(j,i,dir):
    v0 = trees_np[j,i]

    #print(f"j:{j} , i:{i}, treehight:{v0}")

    #dir up=0, down=1, left=2, right=3
    if dir==0:
        opvalue = [-1,0]
    if dir==1: #down
        opvalue = [+1,0]
    if dir==2: # left
        opvalue = [0,-1]
    if dir==3: #right
        opvalue = [0,+1]
    
    score = 0

    nohit=True
    j1,i1 = (j,i)
    
    while nohit:
        j1,i1 = (j1+opvalue[0] , i1+opvalue[1] )
        
        if j1<0 or i1<0 or j1>=trees_np.shape[0] or i1>=trees_np.shape[1]:
            nohit=False
        else:
            score+=1
            v1 = trees_np[j1,i1]
            if v1>=v0:
                nohit=False
    
    return score
```

```python
#Test
print(get_dir_tree_score_at(1,2,0))
print(get_dir_tree_score_at(1,2,1))
print(get_dir_tree_score_at(1,2,2))
print(get_dir_tree_score_at(1,2,3))
```

```python
def get_score_at(j,i):
    return get_dir_tree_score_at(j,i,0)*get_dir_tree_score_at(j,i,1)*get_dir_tree_score_at(j,i,2)*get_dir_tree_score_at(j,i,3)
```

```python
#Test
get_score_at(1,2)
```

```python
get_score_at(3,2)
```

```python
v_get_score_at=np.vectorize(get_score_at)
```

```python
j_mg, i_mg = np.mgrid[0:trees_np.shape[0], 0:trees_np.shape[1]]
```

```python
treescores = v_get_score_at(j_mg, i_mg)
```
