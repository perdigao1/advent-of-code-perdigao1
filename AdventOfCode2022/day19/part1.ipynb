# New Optimizer


Check for how to optimize the production of geode robots

Look at requirements. Compare with materials available, subtract req - current_materials. Any value >0 is a material required. Compare with robots. If the number of robots for that is zero then focus in the production of that robot, by replacing 'adding' the robot requiremens to the required.

Compare again mat_available - required, and compare with robots_available again. Again, if there is a material required and no robot to make, then replace robot with materials

If all robots are there for all materials. Divide the materials required with the number of robots. The highest of the values should indicate the robot that should be prioritised for production.

```python
#This is not working well
def getPriorityRobotPurchase(blueprint, robots_count, materials_count):
    #required is geode = [1,0,0,0]

    production_materials_needed = blueprint['geode']
    print(f"production_materials_needed:{production_materials_needed}, materials_count:{materials_count}")
    required_mat = production_materials_needed - materials_count

    if np.all( required_mat<=0):
        print("Requirements to buy geode already met")
        return 0
    else:
        while True:
            
            print(f"required_mat:{required_mat}")
            
            robots_needed = np.logical_and( required_mat>0 , robots_count==0 )
            print(f"robots_needed:{robots_needed}")

            if np.any(robots_needed):
                #In case there are no robots to produce some of the materials needed
                # then figure out a way to urgently produce these robots
                #TODO
                for i0, value in enumerate(robots_needed):
                    if value:
                        #Sum to the required materials
                        #get key from value in dictionary
                        robot_type= [i for i in ores_dict if ores_dict[i]==i0][0]
                        #print(f"robot_index:{robot_type}")
                        required_mat += blueprint[robot_type]

                        required_mat[i0]=0 #Reset the robots needed
            else:
                break #exit the while loop
        
        #Where there are required materials, divide by the number of robots
        div_res = np.zeros((4))
        np.divide(required_mat, robots_count, where= (required_mat>0) , out=div_res)
        
        #print(f"div_res:{div_res}")
        #Get the maximum in required_mat
        
        max_i = np.argmax(div_res)

        if max_i is None:
            max_i = 2 #Default to buy clay robot if possible

        return max_i
```

```python
def maxAtMinute2(blueprint, robots_count, materials_count, minute, maxminutes):
    #print(f"** {minute} mins, robots_count: {robots_count}, materials_count:{materials_count}")
    global maxgeode_global
    global already_calculated
    
    #Check if this state has already been calculated
    if (*robots_count, *materials_count, minute) in already_calculated:
        return already_calculated[(*robots_count, *materials_count, minute)]

    #Check if there is any chance to get a better result. Number of geode robots and geode materials
    minleft = maxminutes-minute
    maxgeodes_possible = materials_count[0] +1+ (2*(robots_count[0]+1)+minleft)*(minleft+1)/2
    if maxgeodes_possible<maxgeode_global:
        return 0 #Get out if it is not possible to do better

    maxgeodes = 0 #default
    
    
    buylist = getOptionsNext(blueprint, robots_count, materials_count)
    #print(f"buylist:{buylist}")
    
    maxgeode0=0

    for buyrobot0 in buylist:
        materials_count1 = materials_count.copy() #Hopefully, np array copy properly
        robots_count1 = robots_count.copy()
        #print(f"Buying robot {buyrobot0}.")

        if not buyrobot0 is None:
            #Test materials that will be left if we buy the robot
            materials_count1 -= blueprint[buyrobot0]
    
            #print(f"After buying this robot the materials left is {materials_count1}")

        #Update materials based in the number of robots
        materials_count1+= robots_count
        #print(f"Robots collect materials, leading to {materials_count1}")
    
        #Update robot count at end of minute if there was a purchase
        if not buyrobot0 is None:
            robots_count1[buyrobot0]+=1
            #print(f"New robot acquired, robot_count1 is now {robots_count1}")

        minute1 = minute+1
        #Check if reached maximum number of minutes
        if minute1>maxminutes:
            return materials_count1[0]

        #print(f"Entering {minute1} mins...")
        maxgeode0 = maxAtMinute2(blueprint,robots_count1,materials_count1,minute1, maxminutes)
        #print(f"{minute} mins, maxgeode0: {maxgeode0}")

        maxgeodes= max(maxgeodes, maxgeode0)

    if maxgeodes>maxgeode_global:
        #print(f"{minute} min, new geode max: {maxgeodes}")
        maxgeode_global=maxgeodes
        
    already_calculated[(*robots_count, *materials_count, minute)] = maxgeodes

    return maxgeodes
```
