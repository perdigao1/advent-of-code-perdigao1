Code to move elements around in seq_list

```python
def swap(seq_list,i0,i1):
    a = seq_list[i0]
    b = seq_list[i1]
    seq_list[i0]=b
    seq_list[i1]=a
    return seq_list
```

```python
size =len(seq_list)
size
```

```python
def printseq(seq0):
    s=""
    for i,v in seq0:
        s+=str(v)+", "
    print (s)
    return
```

```python
seq_list= seq_list_orig.copy()
for t0 in seq_list_orig:
    #Find location in seq_list
    i0 = seq_list.index( t0 )

    id, v0 = seq_list[i0]

    #print(f"move item ({id} , {v0})")
    #Use value to move it around
    if v0>0:
        #Move right
        j0=i0 #Starting index
        for i in range(v0):
            j1=j0+1
            if j1>=size:
                j1=0
            seq_list=swap(seq_list,j0,j1)
            
            j0=j1 # new index value
    elif v0<0:
        #Move left
        j0=i0 #Starting index
        for i in range(-v0):
            j1=j0-1
            if j1<0:
                j1=size-1
            seq_list=swap(seq_list,j0,j1)
            
            j0=j1 # new index value

    #printseq(seq_list)

print(seq_list)
```
