# Parse map

Convert map to a numpy array of int8

```python
VOID=0
TILE=1
WALL=2
```

```python
conv_dict={' ':VOID , '.':TILE , '#':WALL}
```

need to figure out dimensions of the map

```python
ysize = len(map)
ysize
```

```python
xsize=0
for l0 in map:
    lsize = len(l0)
    xsize=max(xsize,lsize)
xsize
```

Create array with these dimensions and fill with the elements

```python
map_np = np.zeros((ysize,xsize), dtype=np.int8)

for iy, l0 in enumerate(map):
    for ix,char in enumerate(l0):
        map_np[iy,ix] = conv_dict[char]
```

Setup directions

```python
DIR_UP=(-1,0)
DIR_DOWN=(1,0)
DIR_LEFT=(0,-1)
DIR_RIGHT=(0,1)
```

```python
Lrot_dict={DIR_UP:DIR_LEFT , DIR_DOWN:DIR_RIGHT , DIR_LEFT:DIR_DOWN , DIR_RIGHT:DIR_UP} #counterclockwise
Rrot_dict={DIR_UP:DIR_RIGHT , DIR_DOWN:DIR_LEFT , DIR_LEFT:DIR_UP , DIR_RIGHT:DIR_DOWN} #clockwise
```

# Algorithm

```python
#Test
segm = map_np[3,:]
print(np.nonzero( np.logical_and( segm[:-1]==TILE, segm[1:]==VOID) ) )
print(np.nonzero( np.logical_and( segm[:-1]==TILE, segm[1:]==VOID) ) [0].size)
segm = map_np[0,:]
print(np.nonzero( np.logical_and( segm[:-1]==TILE, segm[1:]==VOID) ) )
print(np.nonzero( np.logical_and( segm[:-1]==TILE, segm[1:]==VOID) ) [0].size)
```

```python
def doStep(pos0, dir0):
    #Check against map if one step of the move is possible
    y = pos0[0] #new positons start
    x = pos0[1]

    sizey, sizex = map_np.shape
    isVoid=True

    while isVoid:
        #Move one step
        y+=dir0[0]
        x+=dir0[1]

        if x<0: x=sizex-1
        if x>=sizex: x=0
        if y<0: y=sizey-1
        if y>=sizey: y=0

        #Check if it is wall
        if map_np[y,x] == WALL:
            return None #Cannot move, so stop
        elif map_np[y,x] == VOID:
            isVoid=True
        else:
            isVoid=False

    assert map_np[y,x] == TILE

    return [y,x]
```

```python
#Test
print(doStep( [0,8] , DIR_RIGHT) )
print(doStep( [0,9] , DIR_RIGHT) )
print(doStep( [7,3] , (1,0)) )
```

```python
#Starting position and direction
pos = [ystart, xstart]
dir = DIR_RIGHT
print(f"starting position {pos}, direction:{dir}")
for m0 in moves_list:
    #print(f"Move {m0}")
    if m0.isnumeric():
        #Makes the move N times
        steps = int(m0)
        #print(f"Numeric move, steps {steps}")
        for i in range(steps):
            #print(f"i:{i}")
            res = doStep(pos,dir)
            #print(f"res:{res}")
            if res is None:
                break
            else:
                pos=res
    else:
        #Instruction is a turn
        if m0=='R':
            #Clockwise
            newdir = Rrot_dict[dir]
        else:
            assert m0=='L'
            newdir= Lrot_dict[dir]
        dir=newdir
    #print(f"position after move {m0}: {pos}, direction:{dir}")
```
