```python
#file_input="input_test.txt"
file_input="input.txt"

with open(file_input,'r') as f:
    lines = f.readlines()
```

```python
iline_stackbottom=0
iline_movestart=0

for iline, line0 in enumerate(lines):
    if '1   2' in line0:
        iline_stackbottom=iline-1
        iline_movesstart=iline+2

print(iline_stackbottom)
print(iline_movesstart)
```

```python
ncrates = int(len(lines[0]) / 4)
ncrates
```

```python
#initialise stacks
stacks = []
for istack in range(ncrates):
    stacks.append([])

for iline in range(iline_stackbottom, -1, -1): #Need to end at -1 to include the first line
    line0 = lines[iline]
    stack_letters = [ line0[i] for i in range(1,len(line0),4)]
    
    for i, sl0 in enumerate(stack_letters):
        if not ' ' in sl0:
            stacks[i].append(sl0)

stacks
```

Get moves

```python
#Test using regex
import re

line0 = lines[iline_movesstart]
print(line0)

linecommands = re.findall("\d+", line0)
print(linecommands)
#First is move, second is from, third is to
```

```python
moves_list=[]
for iline in range(iline_movesstart,len(lines)):
    line0 = lines[iline]
    if not line0=='\n':
        move_s = re.findall("\d+", line0)
        move_int = list(map(int, move_s))

        moves_list.append(move_int)
moves_list
```

```python
stacks_end = stacks.copy()

#Execute moves to stacks
for move in moves_list:
    print(move)
    nmoves = move[0]
    for i in range(nmoves):
        st=stacks_end[move[1]-1].pop()
        stacks_end[move[2]-1].append(st)

stacks_end
```

```python
#Top of stack srting

topofstack_string = ''

for stack0 in stacks_end:
    topofstack_string+= stack0[-1]

topofstack_string
```
