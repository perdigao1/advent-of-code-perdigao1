```python
def get_common_letter_in_rucksack(linestring):
    linestring=linestring.strip() # ensure no return characters
    len_2 = int(len(linestring)/2)
    comp0 = linestring[:len_2]
    comp1 = linestring[len_2:]

    #find common character
    commonchar = None
    for char in comp0:
        if char in comp1:
            commonchar=char
    
    print(f"linestring:{linestring}, comp0:{comp0}, comp1:{comp1}, commonchar:{commonchar}")

    return commonchar
```

```python
common_chars=[]

for line0 in lines:
    common_chars.append( get_common_letter_in_rucksack(line0) )

common_chars
```

```python
def get_priority_from_char(char):
    char0=char[0]

    code = ord(char0)

    value=0

    if 65<=code<=90:
        value= code-38
    elif 97<=code<=122:
        value = code-96
    
    return value
```

```python
#rucksack rs
rs_priority_scores = list( map(get_priority_from_char, common_chars) )
rs_priority_scores
```

```python
rs_priority_scores_sum = np.sum(np.array(rs_priority_scores))
rs_priority_scores_sum
```
