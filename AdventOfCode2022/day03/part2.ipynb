```python
from icecream import ic
import numpy as np
```

```python
#file_input="input_test.txt"
file_input="input.txt"
```

```python
with open(file_input,'r') as f:
    lines = f.readlines()
```

```python
#Split rucksacks into groups of 3
lines_grouped_3 = []
for i0 in range(0, len(lines),3):
    group_of_3 = lines[i0:i0+3]
    lines_grouped_3.append(group_of_3)
lines_grouped_3
```

```python
def get_common_letter_in_groups3(group3):
    line0 = group3[0].strip()
    line1 = group3[1].strip()
    line2 = group3[2].strip()

    #find common character between lines
    commonchar = None
    for char in line0:
        if char in line1:
            if char in line2:
                commonchar=char
    
    print(f"commonchar:{commonchar}")

    return commonchar
```

```python
#Test
get_common_letter_in_groups3(lines_grouped_3[0])
```

```python
def get_priority_from_char(char):
    char0=char[0]

    code = ord(char0)

    value=0

    if 65<=code<=90:
        value= code-38
    elif 97<=code<=122:
        value = code-96
    
    return value
```

```python
def get_common_letter_in_groups3_score(group3):
    return get_priority_from_char(get_common_letter_in_groups3(group3))
```

```python
#rucksack rs
rs_group_priority_scores = list( map(get_common_letter_in_groups3_score, lines_grouped_3) )
rs_group_priority_scores
```

```python
rs_priority_scores_sum = np.sum(np.array(rs_group_priority_scores))
rs_priority_scores_sum
```
