```python
import numpy as np
```

```python
file_input = 'input.txt'
#file_input = 'input_test.txt'

with open(file_input,'r') as f:
    lines = f.readlines()
```

# Part1

```python
lines
```

```python
elves_cal=[]

elf0 = []
#Separate calories for each elf
for line0 in lines:
    if line0 != '\n':
        calories=int(line0)
        elf0.append(calories)
    else:
        #print('empty line, next elf')
        elves_cal.append( np.array(elf0) )

        elf0=[] #Resets
if len(elf0)>0:
    elves_cal.append( np.array(elf0) ) #dont forget last line

elves_cal_np = np.array(elves_cal)
```

```python
elves_cal_np
```

```python
elves_sumcal = np.array([np.sum(elf0) for elf0 in elves_cal_np])
```

```python
elves_sumcal
```

```python
elf_max_cal = np.max(elves_sumcal)
elf_max_cal #ok
```

# Part2

Get the three most calories elves

```python
cal_sort = np.sort(elves_sumcal)
cal_sort
```

sum of 3 most

```python
sum_3_most = np.sum(cal_sort[-3:])
sum_3_most
```
