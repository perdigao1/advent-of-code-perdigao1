```python
#file_input = 'input_test.txt'
file_input = 'input.txt'
```

```python
with open(file_input) as f:
    lines = f.read().split('\n')
```

```python
OP_NONE=0
OP_ADD = 1
OP_SUB = 2
OP_MULT = 3
OP_DIV = 4
```

```python
import re
```

```python
#Store information as a dictionary
m_shouting={}
m_operating = {} #store items as tuple monkey : (monkey1, monkey2, operation)
for l0 in lines:
    if l0!='':
        #print(f"l0: {l0}")
        l0_split = l0.split(':')
        monkey = l0_split[0]

        monkey_def = l0_split[1]
        #print(f"monkey_def:{monkey_def}")
        
        if monkey_def.strip().isnumeric():
            m_shouting[monkey] = int(monkey_def)
        else:
            #Check if valid

            x= re.search("[a-z]{4}\s+[\+\-\*\/]\s+[a-z]{4}", monkey_def)
            #print(f"x={x}")
            if x:
                monkeys_op = re.findall("[a-z]{4}", monkey_def)
                monkey1 = monkeys_op[0]
                monkey2 = monkeys_op[1]

                op_char = re.findall("[\+\-\*\/]", monkey_def)
                operation = OP_NONE
                if '+' in op_char:
                    operation = OP_ADD
                elif '-' in op_char:
                    operation = OP_SUB
                elif '*' in op_char:
                    operation = OP_MULT
                elif '/' in op_char:
                    operation = OP_DIV

                assert operation!= OP_NONE, "Error, OP_NONE, parsing didn't work correctly."

                m_tuple = (monkey1, monkey2, operation)
                m_operating[monkey] = m_tuple

            else:
                print(f"details not valid, {monkey_def[1]}")


```

# Calculation algorithm

```python
def getRootValueFor_humn(humn_value):

    m_shouting1 = m_shouting.copy()
    m_operating1 = m_operating.copy()

    m_shouting1['humn'] = humn_value

    i=0

    while len(m_operating1):
        i+=1
        #print(f"iteration: {i}, len(m_operating1): {len(m_operating1)}")
        #m_operating2= m_operating1.copy()
        #Check each operating monkeys individually, and see if can calculate and get the shouting value
        for m0, t0 in list(m_operating1.items()): #Dictionary items changed
            
            m1, m2, op = t0

            #Check if both m1 and m2 are shouting
            if m1 in m_shouting1.keys() and m2 in m_shouting1.keys():
                v1 = m_shouting1[m1]
                v2 = m_shouting1[m2]

                #Do the operation
                value = 0
                if op==OP_ADD:
                    value = v1+v2
                if op==OP_SUB:
                    value = v1-v2
                if op==OP_MULT:
                    value = v1*v2
                if op==OP_DIV:
                    value = v1/v2
                
                #Move this monkey to the shouting list
                m_shouting1[m0]= value

                #Remove it from the m_operating1
                m_operating1.pop(m0)
    
    #Get the 'root' value
    return m_shouting1['root']
```

```python
v_func = np.vectorize(getRootValueFor_humn)
```

Try to optimize using scipy

```python
from scipy import optimize

res = optimize.fsolve(v_func,3.61761395e+12, xtol=1e-12)
res
```
