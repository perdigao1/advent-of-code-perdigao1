make a range being a tuple (start, end)

each assignement being two 'ranges'

Make function that checks complete overlap and count

```python
def check_overlap(range0, range1):
    r0_start, r0_end = range0
    r1_start, r1_end = range1

    bOverlaps= False #default

    start_max = max(r0_start, r1_start)
    end_min = min(r0_end, r1_end)

    if end_min>= start_max:
        bOverlaps=True
    
    return bOverlaps
```

```python
countOverlaps=0
for line0 in lines:
    #get ranges from line
    line0=line0.strip()
    pairs= line0.split(',')

    range0 = list(map(int, pairs[0].split('-')))
    range1 = list(map(int, pairs[1].split('-')))

    #print(f"range0:{range0} , range1:{range1}")

    if check_overlap(range0,range1):
        countOverlaps+=1

countOverlaps
```
