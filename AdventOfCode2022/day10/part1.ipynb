```python
#inputfile='input_test.txt'
inputfile='input.txt'

with open(inputfile,'r') as f:
    lines=f.readlines()
lines = [line0.strip() for line0 in lines]
```

```python
lines
```

```python
addseq = [0,0]
for line0 in lines:
    if 'addx' in line0:
        addseq.append(0)

        v0 = int(line0.split(' ')[1])
        addseq.append(v0)
    elif 'noop' in line0:
        addseq.append(0)
```

```python
addseq
```

```python
Xvalues=[]
X=1
for a in addseq:
    X=X+a
    Xvalues.append(X)
```

```python
check_indexes = (20,60,100,140,180,220)
sum_at_indices = 0
for i0 in check_indexes:
    v0 = Xvalues[i0]
    strength = i0*v0
    print(f"index {i0}, X={v0}, strength = {strength}")
    sum_at_indices += strength

print(f"sum_at_indices={sum_at_indices}")
```
