```python
#inputfile='input_test.txt'
inputfile='input.txt'

with open(inputfile,'r') as f:
    lines=f.readlines()
lines = [line0.strip() for line0 in lines]
```

```python
#Spritepos
X = 1
```

```python
addseq = [0]
for line0 in lines:
    if 'addx' in line0:
        addseq.append(0)

        v0 = int(line0.split(' ')[1])
        addseq.append(v0)
    elif 'noop' in line0:
        addseq.append(0)
```

```python
addseq
```

```python
len(addseq)
```

```python
import numpy as np
pixels_flat = np.full(241, False)
```

```python
Xvalues = []
Xdraw_values = []
#i is the current pixel position
X=0 #sprite position
for i,addvalue in enumerate(addseq):
    X+=addvalue
    Xvalues.append(X)
    irow = i%40
    newpix = False

    xdraw = irow-X 
    Xdraw_values.append(xdraw)
    if 0 <= xdraw <3: 
        newpix= True 
    pixels_flat[i] = newpix
```

```python
import matplotlib.pyplot as plt
plt.plot(Xvalues)
```
