```python
import numpy as np
```

```python
#inputfile='input_test2.txt'
inputfile='input.txt'

with open(inputfile,'r') as f:
    lines=f.readlines()
lines = [line0.strip() for line0 in lines]
```

```python
len(lines)
```

```python
movechar_to_movevector_dict = { 'R':(1,0), 'L':(-1,0), 'U':(0,1), 'D':(0,-1) }
```

```python
# notation (x,y)
startpos = [0,0] #Be carefully when reusing this, make a copy

#Head is the knot at index 0

tailknots = [ startpos.copy() for i in range(10) ]

knot9_track = {tuple(startpos)}
knot9_positions = [startpos.copy()]

knot1_positions=[startpos.copy()]

for line0 in lines:
    line0_split = line0.split(' ')

    #Move head
    hmovedir = line0_split[0]
    hmove_vector = movechar_to_movevector_dict[hmovedir]
    
    hmove_length = int(line0_split[1])
    
    #print(f"line: {line0}, hx,hy={hx},{hy} , tx,ty={tx},{ty}")
    #print(tailknots)

    for ihmoves in range(hmove_length):
        
        #New head position
        tailknots[0][0]+= hmove_vector[0]
        tailknots[0][1]+= hmove_vector[1]
        #head_positions.append( (hx,hy) )
        #print(f"head moved to : {tailknots[0]} ")
        #print(tailknots)

        #print("move knots according to the rules")
        #move knots according to the rules
        for i in range(1,10):
            hx, hy = tailknots[i-1]

            #Move knots, following knot in front (hx,hy)
            #Knots follow according to the rules established by tail
            tx,ty = tailknots[i]

            #print(f"i:{i} , hx:{hx} , hy:{hy} , tx:{tx} , ty:{ty}")

            distx = hx-tx
            disty = hy-ty

            tmovex, tmovey=(0,0)

            if abs(distx)>=2 or abs(disty)>=2:
                #Only move if one of the distances is 2 or over
                
                #Following
                if abs(distx)>=1:
                    tmovex = int(distx / abs(distx))
                if abs(disty)>=1:
                    tmovey = int(disty / abs(disty))
                
                tx+=tmovex
                ty+=tmovey

            tailknots[i] = [tx,ty]
        
        #print(tailknots)
        knot1_positions.append(tailknots[1])
        #Track last elemtn
        knot9_track.add(tuple(tailknots[9]))
        knot9_positions.append(tailknots[9])
```
