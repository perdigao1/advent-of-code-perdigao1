```python
RIGHT=1
CONTINUE=0
WRONG=-1
```

```python
#compare packets
def compare_packets(p0,p1):
    print(f"Compare {p0} vs {p1}")

    if len(p0)==0:
        if len(p1)>0:
            print(f"Left-side run out of items, RIGHT order")
            return RIGHT
        elif len(p1)==0:
            print(f"Empty, Continue")
            return CONTINUE

    #res = CONTINUE # default
    for i,v0 in enumerate(p0):
        if i>=len(p1):
            #Right side run out of items
            print(f"Right-side run out of items, WRONG order")
            return WRONG #Wrong order
        
        #Check item types
        
        v1= p1[i]

        print(f"Compare items {v0} vs {v1}")

        #Both are integers?
        if type(v0)== int and type(v1)==int:
            if v0<v1:
                print(f"{v0} is smaller than {v1}. It is in the RIGHT order")
                return RIGHT #Exit the function now with positive answer
            elif v0>v1:
                print(f"{v0} is larger than {v1}. It is in the WRONG order")
                return WRONG
        
        else:
            res0=CONTINUE
            if type(v0)==int and type(v1)!=int:
                print(f"Mixed types, promoting {v0} to list")
                v0 = [v0]
                
            elif type(v0)!=int and type(v1)==int:
                print(f"Mixed types, promoting {v1} to list")
                v1 = [v1]

            #Now v0 and v1 should be the same type
            res0 = compare_packets(v0,v1)

            if res0== WRONG:
                return WRONG
            elif res0==RIGHT:
                return RIGHT
            
            #res = res0
    
    if len(p0)<len(p1):
        print(f"Left-side run out of items, RIGHT order")
        return RIGHT

    return CONTINUE
    #return res
```

```python
compare_packets(pk0, pk1)
```

```python
sum_index_successes =0
success_indices = []

for ipair, pairs_of_packet0 in enumerate(pairs_of_packets):
    packet0, packet1 = pairs_of_packet0.split('\n')[:2]
    print(f"NEW PAIR packet0:{packet0} , packet1:{packet1}")
    pk0 = json.loads(packet0)
    pk1 = json.loads(packet1)
    res = compare_packets(pk0, pk1)

    if res>=0:
        print("PAIR Result RIGHT")
        success_indices.append(ipair+1)
    else:
        print("PAIR Result: WRONG")
```
