```python
class filedir():
    def __init__(self, name):
        self.name=name
    def getsize(self):
        return 0

class dir(filedir):
    def __init__(self, name, parentdir=None, level=0):
        super().__init__(name)
        self.items=[]
        self.parentdir = parentdir
        self.level=level
    
    def addobj(self, filedirobj):
        level0 = self.level
        self.items.append(filedirobj)
    
    #Overrides
    def getsize(self):
        
        sizes = [it.getsize() for it in self.items]

        sum = np.sum(np.array(sizes))
        return sum

    def __str__(self):
        stabs = ''.join('    '*(self.level+1))
        retstr=''
        retstr+= (f"dir:{self.name} , size:{self.getsize()}, containing...\n")
        for obj in self.items:
            retstr+= stabs+str(obj)+"\n"
        #retstr+= (f"dir:{self.name} end")
        return retstr
    
    def getdir(self,name):
        ret=None
        for it in self.items:
            if it.__class__.__name__=='dir':
                if it.name==name:
                    return it
        return ret


class file(filedir):
    size = 0
    def __init__(self, name, size=0):
        super().__init__(name)
        self.size=size
    def setsize(self,newsize):
        self.size= newsize
    def getsize(self):
        return self.size

    def __str__(self):
        return f"file:{self.name}, size:{self.size}"
```
