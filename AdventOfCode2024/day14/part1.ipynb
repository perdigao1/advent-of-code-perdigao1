# Read file as lines

```python
TEST = False
TEST_SOLUTION = 12

INPUT_FN = "input_test.txt"

if not TEST:
    INPUT_FN = "input.txt"

with open(INPUT_FN) as f:
    lines = [l.rstrip() for l in f]

if lines[-1]=='':
    lines=lines[:,-1]
```

```python
lines
```

```python
from dataclasses import dataclass

@dataclass
class cRobot:
    pos: tuple[int,int]
    vel: tuple[int,int]
```

```python
robots = []

for line0 in lines:
    if len(line0)>0:
        poss , vels = line0.split(" ")
        #print(f"{poss=} , {vels=}")
        posx,posy = map(int,poss.split('=')[1].split(','))
        velx,vely = map(int, vels.split('=')[1].split(','))

        rob0 = cRobot( (posx, posy) , (velx,vely))
        robots.append(rob0)
```

```python
robots
```

```python
robots_final = []

sizex, sizey = 101,103
if TEST:
    sizex, sizey = 11,7
seconds = 100

for rob0 in robots:
    px,py = rob0.pos
    vx,vy = rob0.vel

    posx_final = px+seconds*vx
    posy_final = py+seconds*vy

    posx_final = posx_final % sizex
    posy_final = posy_final % sizey

    robots_final.append(cRobot((posx_final, posy_final), (0,0)))
```

```python
robots_final
```

```python
# Quadrants and multiply
```

```python
import math
count_q_top_left=0
count_q_top_right=0
count_q_bottom_left = 0
count_q_bottom_right = 0
x_mid = sizex / 2
y_mid = sizey / 2

for rob0 in robots_final:
    px,py = rob0.pos
    if px < x_mid-1:
        if py < y_mid-1:
            count_q_top_left+=1
        elif py > y_mid:
            count_q_bottom_left+=1
    elif px > x_mid:
        if py < y_mid-1:
            count_q_top_right+=1
        elif py > y_mid:
            count_q_bottom_right+=1

print(f"{count_q_top_left=}, {count_q_top_right=}, {count_q_bottom_left=}, {count_q_bottom_right=}")
```

```python
safety_factor = count_q_top_left*count_q_top_right*count_q_bottom_left*count_q_bottom_right
safety_factor
```
