```python
#INPUT_FN = "input_test.txt"
INPUT_FN = "input.txt"


with open(INPUT_FN) as f:
    all_file = f.read()
```

```python
all_file
```

```python
import re
```

```python
muls = re.findall("mul\(\d+,\d+\)", all_file)
```

```python
sum_prods = 0
for m in muls:
    numbers = re.findall("\d+", m)
    print(f"{m=}, {numbers=}")
    prod = int(numbers[0])*int(numbers[1])
    sum_prods+=prod
```
