# Read file as lines

```python
TEST = False
TEST_SOLUTION = 480

INPUT_FN = "input_test.txt"

if not TEST:
    INPUT_FN = "input.txt" # 1477762

with open(INPUT_FN) as f:
    lines = [l.rstrip() for l in f]

if lines[-1]=='':
    lines=lines[:,-1]
```

```python
lines
```

```python
from dataclasses import dataclass
@dataclass
class cMachine:
    prize: tuple[int,int]
    button_A: tuple[int,int]
    button_B: tuple[int,int]
```

# Parse file lines

```python
machines=[]

button_A=None
button_B=None
prize = None

for sline in lines:
    if len(sline)>0:
        if "Button A" in sline:
            split0 = sline.split(": ")
            split1 = split0[1].split(",")
            x,y = [ int(chars.split('+')[1]) for chars in split1]
            button_A = (x,y)
        
        elif "Button B" in sline:
            split0 = sline.split(": ")
            split1 = split0[1].split(",")
            x,y = [ int(chars.split('+')[1]) for chars in split1]
            button_B = (x,y)
        elif "Prize" in sline:
            split0 = sline.split(": ")
            split1 = split0[1].split(",")
            x,y = [ int(chars.split('=')[1]) for chars in split1]
            prize = (x,y)

            cmach0 = cMachine(prize, button_A, button_B)
            machines.append(cmach0)
    
```

# Algorithm to calculate tokens

```python
int(np.inner([60.,9.],[3,1]))
```

```python
count_tokens=0

for mach0 in machines:
    # check if buttons do the same, only different lengths
    if mach0.button_A[1]>0 and mach0.button_B[1]>0:
        if mach0.button_A[0]/ mach0.button_A[1] == mach0.button_B[0]/ mach0.button_B[1]:
            print("THE SAME DIRECTION")
    A = np.transpose(np.array( [mach0.button_A, mach0.button_B]))
    B = np.array(mach0.prize)
    X = solve( A ,  B)
    print(f"Machine  ButtA: {mach0.button_A}, ButtB: {mach0.button_B}, Pr: {mach0.prize}")
    if is_all_int(X):
        #print("is_all_int")
        # Check tokens don't exceed 100
        if np.max(X)<=100:
            count_tokens+= np.rint(np.inner(X,[3,1])) #int does what floor does, so it may sometimes give -1 value
            print(f"OK:{X} {count_tokens}")
        else:
            print(f"Too many button presses : {X}")
    else:
        print(f"not all int {X}")
```

# Part 2

Cannot use numpy to solve this as numpy uses int approximations for large numbers

```python
#prize_add = 1 # debug
prize_add = 10000000000000
```

```python
count_tokens=0

for mach0 in machines:
    # check if buttons do the same, only different lengths
    ax,ay = mach0.button_A
    bx,by = mach0.button_B
    tx,ty = mach0.prize

    tx+= prize_add
    ty+= prize_add

    nb = (ax*ty - ay*tx) / (ax*by-ay*bx)
    na = (tx-bx*nb)/ax
    
    print(f"Machine  ButtA: {mach0.button_A}, ButtB: {mach0.button_B}, Pr: {mach0.prize}")
    if na>0 and na.is_integer() and nb>0 and nb.is_integer():
        count_tokens+= int(na)*3 + int(nb)
        print(f"OK:{na, nb} {count_tokens}")
    else:
        print(f"not all int {na,nb}")
```
