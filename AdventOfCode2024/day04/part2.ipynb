```python
#INPUT_FN = "input_test.txt"
INPUT_FN = "input.txt"

with open(INPUT_FN) as f:
    lines = [l.rstrip() for l in f]

if lines[-1]=='':
    lines=lines[:,-1]
```

Look for A's and then look for M's in the diagonals

```python
Xs = []
Ms = []
As = []
Ss = []

for irow, l0 in enumerate(lines):
    for icol, c0 in enumerate(l0):
        if c0=='X': 
            Xs.append( (irow,icol) )
        elif c0=='M':
            Ms.append( (irow, icol) )
        elif c0=='A':
            As.append( (irow,icol))
        elif c0=='S':
            Ss.append((irow,icol))
```

```python
letters_dict = {'X':Xs, 'M':Ms, 'A':As, 'S':Ss}
```

```python
dir_diags = [(1,1), (1,-1), (-1,1), (-1,-1)]
```

```python
count=0

for a0 in As:

    # Check along diagonals if there are any M's and opposite S's
    count_mas = 0
    for dd0 in dir_diags:
        test_M_pos = ( a0[0]+dd0[0], a0[1]+dd0[1])
        if test_M_pos in Ms:
            # Check for S on the other side
            test_S_pos = ( a0[0]-dd0[0], a0[1]-dd0[1])
            if test_S_pos in Ss:
                count_mas+=1
    if count_mas==2:
        print(f"X-MAS cross at center location {a0}")
        count+=1
```
