I will implement as a linked list

```python
class cStone():
    val_str = None
    next = None
    #prev = None

    def __init__(self, val= None):
        self.val_str = val
```

Parse input

```python
split_input = LINE_INPUT.strip().split()
split_input
```

```python
stones=[]

assert len(split_input)>1

for s in split_input:
    stones.append( cStone(s) )

#sort links
for i,cst in enumerate(stones):
    # # previous
    # if i>0:
    #     cst.prev = stones[i-1]

    #next
    if i<len(stones)-1:
        cst.next = stones[i+1]
```

```python
start_stone = stones[0]
start_stone.val_str
```

```python
def print_stones(stone0):
    text=""
    s=stone0
    while s:
        text+= f"{s.val_str} "
        s=s.next
    text=text[:-1] #remove last space
    print(text)
```

```python
print_stones(start_stone)
```

Run the "blinking"

```python
cur_stone = start_stone

for iteration in range(25):
    cur_stone= start_stone

    while cur_stone is not None:
        #print(f"{cur_stone.val_str=}")
        next_it_stone = cur_stone.next
        str0 = cur_stone.val_str

        if str0 == "0":
            cur_stone.val_str = "1"
        elif len(str0)%2==0: # even
            #print(f"{str0=} is even digits")
            new_cur_stone_val = str(int(str0[:len(str0)//2]))
            cur_stone.val_str = new_cur_stone_val
            #print(f"old stone new value {cur_stone.val_str}")
            
            new_stone_value = str(int(str0[len(str0)//2: ]))
            st2 = cStone(new_stone_value)
            #print(f"new stone value {new_stone_value}")
            cur_stone.next = st2
            st2.next = next_it_stone
        else:
            cur_stone.val_str = str(int(cur_stone.val_str)*2024)

        cur_stone = next_it_stone
    
    print(f"blink {iteration+1}")
    print_stones(start_stone)

```

```python
print_stones(start_stone)
```

```python
def count_number_of_stones(start_stone):
    count=0
    s=start_stone
    while s:
        count+=1
        s=s.next
    return count
```
