```python
#INPUT_FN = "input_test.txt" #2
INPUT_FN = "input.txt"

with open(INPUT_FN) as f:
    lines = [l.rstrip() for l in f]

if lines[-1]=='':
    lines=lines[:,-1]
```

```python
def is_diff_ok(d1,d2):
    diff = abs(d1-d2)
    if diff==0 or diff>3:
        return False
    return True
```

```python
safe_count = 0

for report in lines:
    values = list(map(int,report.split(' ') ))
    is_safe = True #default

    cause=""
        
    first_diff= values[1]-values[0]

    if not is_diff_ok(values[1],values[0]):
        cause="First values are either equal or diff>3"
        is_safe=False
    else:
        for i in range(2, len(values)):
            val_im1, val_i = values[i-1], values[i]
            if not is_diff_ok(val_i, val_im1):
                cause = "diff=0 or >3, not valid"
                is_safe=False
                break
            if (val_i-val_im1)*first_diff<0:
                cause = "change of incr-decr direction"
                is_safe=False
                break
    
    print(f"{values=}, {is_safe=}, {cause=}")
    if is_safe:
        safe_count+=1

print(safe_count)
```
