# Parse map

Wall positions, boulders and robot position represented in list of tuples

```python
walls=[]
boulders=[]
robot_pos0 = None

max_row = 0
max_col = 0

for irow,s in enumerate(s_map):
    for icol, c in enumerate(s):
        pos0 = (irow,icol*2) # times 2, because each item covers 2 spaces
        match c:
            case '#':
                walls.append( pos0 )
                walls.append( (irow,icol*2+1))
                max_row = max(pos0[0], max_row)
                max_col = max(pos0[1]+1, max_col)
            case 'O':
                boulders.append( pos0 )
            case '@':
                robot_pos0 = pos0
```

```python
def print_map(walls0, boulders0, rpos0):
    global max_row, max_col

    smap0 = []
    for i in range(max_row+1):
        smap0.append( ['.']* (max_col+1) )
    
    print(f"{len(smap0)=}, {len(smap0[0])=}")
    
    for r,c in walls0:
        smap0[r][c] = '#'
    
    for r,c in boulders0:
        smap0[r][c:c+2] = '[]'
    
    smap0[rpos0[0]][rpos0[1]]='@'

    for s in smap0:
        #print(s)
        print ( ''.join(s))

```

# Calculation algorithm

walls is fixed.

* Iterate though movements
* for each movement check boulder in a recursive fashion. function that returns true if can move in that direction
* boulders and walls need a special algorithm detect and move. There can be a situation that one boulder moves two other boulders

```python
# initialise
robot_pos_cur = robot_pos0
boulders_cur = boulders.copy()
WALLS = walls

# def can_move_to(try_pos, dir, boulders0):
#     global WALLS
#     if try_pos in WALLS:
#         return False
    
#     if try_pos in boulders0:
#         # check if boulder can be moved

#         try_pos0 = ( try_pos[0] + dir[0] , try_pos[1] + dir[1])

#         res =  can_move_to(try_pos0, dir, boulders0)
#         if res:
#             # move boulder
#             boulders0.remove( try_pos)
#             boulders0.append( try_pos0 )
#             return True
        
#         return False
    
#     return True


def can_move_to(try_pos, dir, boulders0):
    global WALLS

    if not isinstance(try_pos, list):
        # need to check all positions
        try_pos = [try_pos]
    
    #can_move= True # default

    boulders_hit = set()
    for tpos0 in try_pos:

        if tpos0 in WALLS:
            # can_move=False
            # break
            return False
        
        if tpos0 in boulders0:
            boulders_hit.add(tpos0)

        tpos1 = (tpos0[0],tpos0[1]-1)
        if tpos1 in boulders0:
            boulders_hit.add(tpos1)
        
    if len(boulders_hit)==0:
        printd("No boulders hit. ok to move")
        return True
    else:
        printd(f"boulders hit {boulders_hit}")
        # create a new list of positions to check
        check_next_positions = set()

        for b0 in boulders_hit:
            if abs(dir[0])==1: #vertical movement
                check_next_positions.add( ( b0[0] + dir[0] , b0[1]) )
                check_next_positions.add( ( b0[0] + dir[0] , b0[1]+1) )
            else: # horizontal
                if dir[1]==-1:
                    check_next_positions.add( ( b0[0] , b0[1]-1) )
                else:
                    check_next_positions.add( ( b0[0] , b0[1]+2) )
        printd(f"{check_next_positions=}")
        res =  can_move_to(list(check_next_positions), dir, boulders0)

        if res:
            # move boulders
            for b0 in boulders_hit:
                new_b_pos = ( b0[0]+dir[0] , b0[1]+dir[1] )
                boulders0.remove( b0 )
                boulders0.append( new_b_pos )
            return True
    
    return False

for imoves,c in enumerate(s_moves):
    dir = None
    printd(f"Move {c}")
    match c:
        case '<':
            dir = (0,-1)
        case '>':
            dir = (0,1)
        case '^':
            dir = (-1,0)
        case 'v':
            dir = (1,0)
        
    if dir is not None:
        try_pos = ( robot_pos_cur[0] + dir[0] , robot_pos_cur[1] + dir[1])
        printd(f"move by {dir=}, {try_pos=}")
        #print(f"{try_pos=}, {dir=}, {boulders_cur=}")
        res = can_move_to(try_pos, dir, boulders_cur)
        if res:
            robot_pos_cur = try_pos
    if TEST:
        print_map(WALLS, boulders_cur, robot_pos_cur)
```
