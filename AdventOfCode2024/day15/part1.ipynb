# Parse map

Wall positions, boulders and robot position represented in list of tuples

```python
walls=[]
boulders=[]
robot_pos0 = None

max_row = 0
max_col = 0

for irow,s in enumerate(s_map):
    for icol, c in enumerate(s):
        pos0 = (irow,icol)
        match c:
            case '#':
                walls.append( pos0 )
                max_row = max(pos0[0], max_row)
                max_col = max(pos0[1], max_col)
            case 'O':
                boulders.append( pos0 )
            case '@':
                robot_pos0 = pos0
```

# Calculation algorithm

walls is fixed.

* Iterate though movements
* for each movement check boulder in a recursive fashion. function that returns true if can move in that direction

```python
def printd(s):
    if TEST:
        print(s)
```

```python
def print_map(walls0, boulders0, rpos0):
    global max_row, max_col

    smap0 = []
    for i in range(max_row+1):
        smap0.append( ['.']* (max_col+1) )
    
    print(f"{len(smap0)=}, {len(smap0[0])=}")
    
    for r,c in walls0:
        smap0[r][c] = '#'
    
    for r,c in boulders0:
        smap0[r][c] = 'O'
    
    smap0[rpos0[0]][rpos0[1]]='@'

    for s in smap0:
        #print(s)
        print ( ''.join(s))

```

```python
# initialise
robot_pos_cur = robot_pos0
boulders_cur = boulders.copy()
WALLS = walls

def can_move_to(try_pos, dir, boulders0):
    global WALLS
    if try_pos in WALLS:
        return False
    
    if try_pos in boulders0:
        # check if boulder can be moved

        try_pos0 = ( try_pos[0] + dir[0] , try_pos[1] + dir[1])

        res =  can_move_to(try_pos0, dir, boulders0)
        if res:
            # move boulder
            boulders0.remove( try_pos)
            boulders0.append( try_pos0 )
            return True
        
        return False
    
    return True

for imoves,c in enumerate(s_moves):
    dir = None
    match c:
        case '<':
            dir = (0,-1)
        case '>':
            dir = (0,1)
        case '^':
            dir = (-1,0)
        case 'v':
            dir = (1,0)
        
    if dir is not None:
        try_pos = ( robot_pos_cur[0] + dir[0] , robot_pos_cur[1] + dir[1])
        printd(f"move by {dir=}, {try_pos=}")
        #print(f"{try_pos=}, {dir=}, {boulders_cur=}")
        res = can_move_to(try_pos, dir, boulders_cur)
        if res:
            robot_pos_cur = try_pos
    if TEST:
        print("Move ", c)
        print_map(WALLS, boulders_cur, robot_pos_cur)
```
