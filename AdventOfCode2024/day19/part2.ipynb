```python
TEST = False

INPUT_FN = "input_test.txt"
TEST_SOLUTION = 16

if not TEST:
    INPUT_FN = "input.txt"
```

```python
with open(INPUT_FN) as f:
    s = f.read()

patterns_block, designs_block = s.split('\n\n')
```

Parse

```python
patterns_block
```

```python
patterns = [s.strip() for s in patterns_block.split(',') ]
patterns
```

```python
designs = designs_block.split('\n')
if designs[-1] == "":
    designs = designs[:-1]
designs
```

Need to find how many designs are possible with the available patterns

run DFS search with memoisation

```python
from functools import lru_cache

@lru_cache
def count_possible(des0:str):
    global patterns
    pats0 = patterns
    # workout whether the design is possible with any of the patterns given
    if len(des0)==0:
        return 1 #whole pattern ok
    
    count=0

    for p0 in pats0:
        if len(p0)>len(des0):
            continue # skip
        if p0 == des0[:len(p0)]:
            # pattern matches. Check remaining design
            des1 = des0[len(p0):]
            pats1 = pats0.copy()

            res = count_possible(des1)
            count+=res
    
    return count
```

```python
count_all_valid=0
for des0 in designs:
    #print(f"Design: {des0}")
    res = count_possible(des0)
    print(f"Design: {des0}, count valid: {res}")
    count_all_valid+= res
```

```python
if TEST:
    assert count_all_valid==TEST_SOLUTION
```
