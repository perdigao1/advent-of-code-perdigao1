# Day 18 - RAM Run

```python
TEST = False

INPUT_FN = "input_test.txt"
SIZE = 7 # 0 to 6
NBYTESFALL = 12
TEST_SOLUTION = 22

if not TEST:
    INPUT_FN = "input.txt"
    SIZE=71 # 0 to 70
    NBYTESFALL = 1024
```

```python
with open(INPUT_FN) as f:
    lines = [l.rstrip() for l in f]

if lines[-1]=='':
    lines=lines[:,-1]
```

```python
lines
```

Create a graph using networkx

```python
from networkx import Graph
```

```python
byte_poss = []
for l0 in lines:
    spl = l0.split(',')
    e0 = int(spl[0])
    e1 = int(spl[1])
    byte_poss.append((e0,e1))
byte_poss
```

```python
bytes_fallen = byte_poss[:NBYTESFALL]
bytes_fallen
```

```python
# draw byte map
for irow in range(SIZE):
    row_txt_list = ['.']*SIZE
    for icol in range(SIZE):
        if (icol, irow) in bytes_fallen:
            row_txt_list[icol] = '#'
    
    print( "".join(row_txt_list) )
```

```python
edges = []
for ix in range(SIZE):
    for iy in range(SIZE):
        trypos0 = (ix,iy)
        if trypos0 not in bytes_fallen:
            #check right
            if ix < SIZE-1:
                if (ix+1,iy) not in bytes_fallen:
                    edges.append( (trypos0 , (ix+1, iy)))
            if iy < SIZE-1:
                if (ix,iy+1) not in bytes_fallen:
                    edges.append( (trypos0 , (ix, iy+1)) )
```

```python
edges
```

```python
G = Graph(edges)
```

```python
from networkx import dijkstra_path
```

```python
help(dijkstra_path)
```

```python
path0 = dijkstra_path( G, (0,0), (SIZE-1, SIZE-1))
path0
```

```python
steps = len(path0)-1
```

```python
if TEST:
    assert steps== TEST_SOLUTION
```

```python
print("Solution: ", steps)
```
