```python
TEST = False

INPUT_FN = "input_test.txt"
TEST_SOLUTION = 6

if not TEST:
    INPUT_FN = "input.txt"
```

```python
with open(INPUT_FN) as f:
    s = f.read()

patterns_block, designs_block = s.split('\n\n')
```

```python
patterns = [s.strip() for s in patterns_block.split(',') ]
patterns
```

```python
designs = designs_block.split('\n')
if designs[-1] == "":
    designs = designs[:-1]
designs
```

Need to find how many designs are possible with the available patterns

I will try a DFS search

```python
def dfs_is_possible(des0:str, pats0: list, depth0, list_pats = [] ):
    # workout whether the design is possible with any of the patterns given
    #print(f"{depth0=}, {des0=} , {pats0=}")
    if len(des0)==0:
        print (f"list patterns: {list_pats}")
        print(f"result {"".join(list_pats)}")
        
        return True #whole pattern ok
    
    
    for p0 in pats0:
        if len(p0)>len(des0):
            continue # skip
        if p0 == des0[:len(p0)]:
            # pattern matches. Check remaining design
            des1 = des0[len(p0):]
            pats1 = pats0.copy()
            #pats1.remove(p0) # can do repeating patterns
            list_pats1 = list_pats.copy()
            list_pats1.append(p0)

            res = dfs_is_possible(des1, pats1, depth0+1, list_pats1)
            if res:
                return True
    
    return False
```

```python
count_valid=0
for des0 in designs:
    print(f"Design: {des0}")
    res = dfs_is_possible(des0, patterns,0)
    if res:
        count_valid+=1
        print("Design IS possible")
    else:
        print("Design not possible")
```

```python
if TEST:
    assert count_valid==TEST_SOLUTION
```
