# Read file as lines

```python
#INPUT_FN = "input_test.txt" #1928
INPUT_FN = "input.txt"

with open(INPUT_FN) as f:
    lines = [l.rstrip() for l in f]

if lines[-1]=='':
    lines=lines[:,-1]
```

```python
assert len(lines)==1
```

We can now create very long lisr with the structure of the drive

```python
drive_list_orig = []

id=0

for i,c in enumerate(lines[0]):
    if i%2==0:
        #file
        drive_list_orig += [id]*int(c)
        id+=1
    else:
        drive_list_orig += [None]*int(c)
```

# Defrag

use double pointer

```python
p0 = 0
p1 = len(drive_list_orig)-1

drive_list = drive_list_orig.copy()

while True:
    # find empty loc
    while drive_list[p0] is not None and p0<len(drive_list):
        p0+=1
    
    if p0>=p1 or p0>= len(drive_list)-1:
        print("reached second pointer or end, leaving")
        break

    print (f"empty location {p0=}")

    # find digit to move to empty location
    while drive_list[p1] is None and p1>p0:
        p1-=1
    
    print (f"found digit to move at {p1=}")

    #move digit
    drive_list[p0] = drive_list[p1]
    
    p0+=1
    p1-=1
```

```python
drive_list0 = drive_list[:p1+1]
if drive_list0[-1] is None:

    drive_list0 = drive_list0[:-1]
```

# checksum

```python
checksum = 0
for i, v in enumerate(drive_list0):
    checksum+= i*v
checksum
```
