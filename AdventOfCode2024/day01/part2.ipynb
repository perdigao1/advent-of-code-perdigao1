```python
# INPUT_FN = "input_test.txt" #31
INPUT_FN = "input.txt"

with open(INPUT_FN) as f:
    lines = [l.rstrip() for l in f]

if lines[-1]=='':
    lines=lines[:,-1]
```

```python
lines
```

with the second list, we need to count how much each of the values appear. Create a counter dictionary

```python
col1_vals = []
#col2_vals = []
col2_dict = {}

for line in lines:
    split0 = line.split("   ")
    #print(split0)
    val1, val2 = map(int, split0)
    #print(f"{val1=}, {val2=}")
    col1_vals.append(val1)
    
    if val2 in col2_dict.keys():
        col2_dict[val2]+=1
    else:
        col2_dict[val2]=1 #create

print(col1_vals)
print(col2_dict)
```

Get similarity score

```python
sim_sum=0

for val1 in col1_vals:
    sim_score = val1 * col2_dict.get(val1, 0)
    sim_sum += sim_score
print(sim_sum)
```
