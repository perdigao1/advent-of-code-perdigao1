```python
#INPUT_FN = "input_test.txt" #11
INPUT_FN = "input.txt"

with open(INPUT_FN) as f:
    lines = [l.rstrip() for l in f]

if lines[-1]=='':
    lines=lines[:,-1]
```

```python
col1_vals = []
col2_vals = []

for line in lines:
    split0 = line.split("   ")
    #print(split0)
    val1, val2 = map(int, split0)
    #print(f"{val1=}, {val2=}")
    col1_vals.append(val1)
    col2_vals.append(val2)

print(col1_vals)
print(col2_vals)
```

```python
col1_sort = sorted(col1_vals)
col2_sort = sorted(col2_vals)

print(col1_sort)
print(col2_sort)
```

```python
sum_dists=0
for i in range(len(col1_sort)):
    dist = abs(col1_sort[i]-col2_sort[i])
    sum_dists+=dist
print(sum_dists)
```
