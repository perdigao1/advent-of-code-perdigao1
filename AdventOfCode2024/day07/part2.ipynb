# Load file and parse

```python
#INPUT_FN = "input_test.txt"
INPUT_FN = "input.txt"

with open(INPUT_FN) as f:
    lines = [l.rstrip() for l in f]

if lines[-1]=='':
    lines=lines[:,-1]
```

Get the first number on each line and then the others successively

```python
line_data = []

for line0 in lines:
    first_str, rest_str = line0.split(":")

    first_number = int(first_str)

    numbers = list(map(int, rest_str.strip().split()))

    line_data.append( (first_number, numbers))
```

For testing we can explore all possibilities by using binary. Zero for +, 1 for mult.

Explore all operators combinations from base3 000 to 222

```python
target, numbers = line0
n_ops = len(numbers)-1
print(f"{n_ops=}")
#for i in range(2**n_ops):
for i in range(3**n_ops):
    i_str3 = np.base_repr(i,3,n_ops)
    i_str3 = i_str3[-n_ops:]
    result = numbers[0] # initialize
    print(f"{i=} , {i_str3=}")
    #iterate operations
    for i_s, s in enumerate(i_str3):
        next_num = numbers[i_s+1]
        if s=='0':
            result *= next_num
        elif s =='1':
            result += next_num
        else:
            #combine digits
            print("combine")
            result = int(str(result)+str(next_num))
    print (f"{i=} , {result=}")
    if result == target:
        print("Target result found")
        break
```

It appears to work

Do for all of them and sum the results that work

```python
true_results_sum = 0

for i_line, line0 in enumerate(line_data):
    print(f"{i_line=}: {line0}")
    target, numbers = line0
    n_ops = len(numbers)-1
    #print(f"{n_ops=}")
    #for i in range(2**n_ops):
    for i in range(3**n_ops):
        i_str3 = np.base_repr(i,3,n_ops)
        i_str3 = i_str3[-n_ops:]
        result = numbers[0] # initialize
        #print(f"{i=} , {i_str3=}")
        #iterate operations
        for i_s, s in enumerate(i_str3):
            next_num = numbers[i_s+1]
            if s=='0':
                result *= next_num
            elif s =='1':
                result += next_num
            else:
                #combine digits
                #print("combine")
                result = int(str(result)+str(next_num))
        #print (f"{i=} , {result=}")
        if result == target:
            print("Target result found")
            true_results_sum+= target
            break
            

print(f"{true_results_sum=}")
```
