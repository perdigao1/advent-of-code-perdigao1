```python
#INPUT_FN = "input_test.txt" #123
INPUT_FN = "input.txt"

with open(INPUT_FN) as f:
    lines = [l.rstrip() for l in f]

if lines[-1]=='':
    lines=lines[:,-1]
```

```python
lines
```

```python
rules= []
updates = []

at_rules=True
for l0 in lines:
    if l0=="":
        at_rules=False
        continue #skip to next line
    if at_rules:
        values = list(map(int, l0.split('|')))
        rules.append(values)
    else:
        values = list(map(int, l0.split(',')))
        updates.append(values)
```

```python
rules
```

```python
def correctify_update(up0):
    up_corrected = up0.copy()

    was_incorrect = False
    has_changed = True
    while has_changed:
        has_changed=False

        # Check all rules one-by-one
        for i0,i1 in rules:
            if i0 in up_corrected and i1 in up_corrected:
                loc0 = up_corrected.index(i0)
                loc1 = up_corrected.index(i1)
                if loc1<loc0:
                    #print(f"Wrong placement {loc0}, {loc1}")
                    # swap elements
                    a,b = up_corrected[loc0], up_corrected[loc1]
                    up_corrected[loc0], up_corrected[loc1] = b,a
                    has_changed=True
                    was_incorrect=True
    
    return up_corrected, was_incorrect
```

```python
score = 0

for up0 in updates:
    upd_corr, was_incorrect = correctify_update(up0)
    print(f"{up0=}, Corrected: {upd_corr}")
    
    assert len(upd_corr)== len(up0)
    if was_incorrect:
        midv = upd_corr[ len(upd_corr)//2 ]
        score+=midv
```
