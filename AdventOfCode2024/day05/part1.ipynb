```python
#INPUT_FN = "input_test.txt" #143
INPUT_FN = "input.txt"

with open(INPUT_FN) as f:
    lines = [l.rstrip() for l in f]

if lines[-1]=='':
    lines=lines[:,-1]
```

```python
lines
```

```python
rules= []
updates = []

at_rules=True
for l0 in lines:
    if l0=="":
        at_rules=False
        continue #skip to next line
    if at_rules:
        values = list(map(int, l0.split('|')))
        rules.append(values)
    else:
        values = list(map(int, l0.split(',')))
        updates.append(values)
```

```python
rules
```

```python
def is_update_correct(up0):

    # Check all rules one-by-one
    for i0,i1 in rules:
        if i0 in up0 and i1 in up0:
            loc0 = up0.index(i0)
            loc1 = up0.index(i1)
            if loc1<loc0:
                return False
    
    return True
```

```python
score = 0

for up0 in updates:
    if is_update_correct(up0):
        # Get middle value
        midv = up0[ len(up0)//2 ]
        score+=midv
```
