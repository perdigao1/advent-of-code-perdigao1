Cards, poker. Need to rank them

```python
#INPUT_FN = "input_test.txt"
INPUT_FN = "input.txt"
```

```python
with open(INPUT_FN) as f:
    lines = [l.rstrip() for l in f]
```

```python
cards_ord_by_value = ['J','2','3','4','5','6','7','8','9','T','Q','K','A']

cards_to_value_dict = dict()

for i in range(len(cards_ord_by_value)):
    cards_to_value_dict[cards_ord_by_value[i]] = i+2
cards_to_value_dict
```

```python
HIGHCARD,ONEPAIR,TWOPAIR,THREEKIND,FULLHOUSE,FOURKIND,FIVEKIND = 1,2,3,4,5,6,7
```

```python
from collections import Counter

class CardHand():
    def __init__(self, line:str):
        items = line.split(" ")
        if len(items)==2:
            self.hand = items[0]
            self.bid = int(items[1])

            self.type=0

            #check all distinct
            unique0 = Counter(self.hand)

            if 'J' in unique0.keys():
                if unique0['J']!=5:
                    #Promote
                    nj = unique0.pop('J')
                
                    tuple_counts_asc = sorted(unique0.items() , key = lambda x: x[1], reverse=True )#sort dict by counts descending

                    #Promote best to better
                    el0 = tuple_counts_asc[0]
                    new_dict_entry = ( el0[0], el0[1]+nj)
                    
                    tuple_counts_asc[0] = new_dict_entry
                    #print(tuple_counts_asc)

                    unique0 = dict(tuple_counts_asc)

            if len(unique0.keys())==5:
                self.type = HIGHCARD
                return
            if len(unique0.keys())==4:
                #there must be one pair
                self.type = ONEPAIR
                return
            if len(unique0.keys())==3:
                # can be two pair or 3 of a kind
                if 3 in unique0.values():
                    self.type = THREEKIND
                else:
                    self.type = TWOPAIR
                return
            if len(unique0.keys())==2:
                if 4 in unique0.values():
                    self.type = FOURKIND
                else:
                    self.type = FULLHOUSE
                return
            if len(unique0.keys())==1:
                self.type = FIVEKIND
                return
        print(f"ERROR with {line}")
    
    def __lt__(self,other):
        if self.type == other.type :
            for i in range(5):
                v_self = cards_to_value_dict[self.hand[i]]
                v_other = cards_to_value_dict[other.hand[i]]
                if v_self != v_other:
                    return v_self < v_other
                
        else:
            return self.type < other.type
        
```

parse card hands

```python
card_hands = []

for l0 in lines:
    card_hands.append( CardHand(l0) )
```

```python
for c1 in card_hands:
    print (f"hand:{c1.hand} , type: {c1.type} , bid:{c1.bid}")
```

sort

```python
card_hands_sorted = sorted(card_hands)
```

```python
for c1 in card_hands_sorted:
    print (f"hand:{c1.hand} , type: {c1.type} , bid:{c1.bid}")
```

```python
res_sum = 0
for i, c2 in enumerate(card_hands_sorted):
    print(f"{c2.bid} * {(i+1)}")
    res_sum += (i+1)*c2.bid
res_sum
```
