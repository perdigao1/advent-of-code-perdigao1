```python
import numpy as np
import re
```

```python
#with open("input_test.txt", 'r') as f0:
with open("input.txt", 'r') as f0:
    lines0= f0.readlines()

lines = [l0.strip() for l0 in lines0]
```

```python
lines
```

```python
# Convert lines to array

dimx =  len(lines[0])
dimy = len(lines)

digits = np.zeros((dimy,dimx), dtype=bool)
symbols = np.zeros_like(digits, dtype=bool)

for iy,l0 in enumerate(lines):
    for ix,s0 in enumerate(l0):
        if s0.isdigit():
            digits[iy,ix]=True
        elif s0!='.':
            # is symbol then
            symbols[iy,ix]=True
```

```python
#Iterate to mark the part numbers
# Use convolutions
from scipy import ndimage
conv_kern = np.ones((3,3), dtype=np.uint8)

symbols_i = np.array(symbols).astype(np.uint8) # make a copy

symbols_conv= ndimage.convolve(symbols_i, conv_kern, mode='constant', cval=0)

symbols_conv_bool = symbols_conv>0

digits_to_account = symbols_conv_bool *  digits #Boolean and
```

```python
#Iterate to mark the part numbers
# Use convolutions
from scipy import ndimage
conv_kern = np.ones((3,3), dtype=np.uint8)

symbols_i = np.array(symbols).astype(np.uint8) # make a copy

ndigits=0
i=0
while True:
    print(i)
    symbols_conv= ndimage.convolve(symbols_i, conv_kern, mode='constant', cval=0)
    symbols_conv_bool = symbols_conv>0
    symbols_i = symbols_conv_bool *  digits #Boolean and
    ndigits1 = np.sum(symbols_i)
    if ndigits1<= ndigits:
        break
    ndigits = ndigits1
    i+=1
```

```python
#Get the values

values=[]

for iy,l0 in enumerate(lines):
    filt_line= list(l0)
    for ix,s0 in enumerate(l0):
        if not symbols_i[iy,ix]:
            filt_line[ix]=' '
    filt_line="".join(filt_line)
    print(f"iy:{iy} , filt_line:{filt_line}")

    r0 = re.findall("[0-9]+",filt_line)

    for s in r0:
        values.append(int(s))
```
