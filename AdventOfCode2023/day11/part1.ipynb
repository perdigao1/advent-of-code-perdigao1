```python
#INPUT_FN = "input_test.txt" # 374

INPUT_FN = "input.txt"

with open(INPUT_FN) as f:
    lines = [l.rstrip() for l in f]

if lines[-1]=='':
    lines=lines[:,-1]
```

```python
lines
```

Use numpy and convert galaxies to numbers

```python
nrows = len(lines)
ncols = len(lines[0])
print(f"{nrows=}, {ncols=}")
```

```python
import numpy as np
univ = np.zeros((nrows, ncols), dtype=np.uint32)
gcount=1
for irow,line in enumerate(lines):
    for icol,s in enumerate(line):
        if s=='#':
            univ[irow,icol]=gcount
            gcount+=1
```

```python
univ
```

Find empty rows and cols and expands

```python
empty_rows = np.any(univ>0, axis=1)
empty_rows
```

Expands where it has no galaxies

```python
irows_no_galaxies = np.nonzero(empty_rows==False)[0]
irows_no_galaxies
```

```python
expand_row = np.zeros((univ.shape[1]), dtype=np.uint32)
expand_row
```

```python
univ_row_exp = np.insert(univ, irows_no_galaxies, expand_row, axis=0)
```

```python
univ_row_exp
```

column-wise

```python
empty_cols = np.any(univ_row_exp>0, axis=0)
empty_cols
```

```python
icols_no_galaxies = np.nonzero(empty_cols==False)[0]
icols_no_galaxies
```

```python
expand_col = np.zeros((univ_row_exp.shape[0],1), dtype=np.uint32)
expand_col
```

```python
univ_row_exp.shape
```

```python
univ_row_col_exp = np.insert(univ_row_exp, icols_no_galaxies, expand_col, axis=1)
```

```python
univ_row_col_exp
```

OK.

Now we need to get manhatan distances between combinations

```python
n_galaxies = univ_row_col_exp.max()
n_galaxies
```

```python
# Try between galaxy 1 and 7

g1 = np.where(univ_row_col_exp==1)
g1 = (g1[0][0], g1[1][0])
g1
```

```python
def get_coords_of_galaxy(N):
    gN = np.where(univ_row_col_exp==N)
    res = (gN[0][0], gN[1][0])
    return res
```

```python
get_coords_of_galaxy(7)
```

```python
def manh_dist(c1, c2):
    dist = abs(c1[0]-c2[0]) + abs(c1[1]-c2[1])
    return dist
```

```python
manh_dist( get_coords_of_galaxy(1), get_coords_of_galaxy(7))
```

```python
manh_dist( get_coords_of_galaxy(3), get_coords_of_galaxy(6))
```

Do all combinations of galaxies

```python
import itertools

g_combs = list(itertools.combinations(range(1,n_galaxies+1),2))
g_combs
```

```python
dist_sum = 0

for comb1 in g_combs:
    dist_sum+= manh_dist( get_coords_of_galaxy(comb1[0]), get_coords_of_galaxy(comb1[1]))
dist_sum
```
