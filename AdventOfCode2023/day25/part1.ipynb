```python
#INPUT_FN = "input_test.txt"
INPUT_FN = "input.txt"

with open(INPUT_FN) as f:
    lines = [l.rstrip() for l in f]
```

Try use networkx and visualise

```python
import networkx
```

```python
G = networkx.Graph()
```

Grow by adding edges

```python
for line in lines:
    if len(lines)>0:
        sleft, sright = line.split(':')

        print(f"sleft:{sleft} , sright:{sright}")
        #nleft = sleft.replace(' ', '')

        sright = sright.removeprefix(' ')
        nodes_right = sright.split(' ')

        for nr0 in nodes_right:
            if nr0=='':
                print(f"error nr0:{nr0}")
            G.add_edge(sleft, nr0)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(20,20))

networkx.draw_spring(G,with_labels=True, font_size=9)
```

ccp-fvm , llm-lhg, frl-thx

Split these edges, then run connected components

```python
G.remove_edges_from([('ccp','fvm'), ('llm','lhg') ,('frl','thx')])
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(20,20))

networkx.draw_spring(G,with_labels=True, font_size=9)
```

```python
res = networkx.connected_components(G)
```

```python
res
```

```python
subgraphs = [G.subgraph(c).copy() for c in res]
```
