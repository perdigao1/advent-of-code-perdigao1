```python
import numpy as np
import re

#INPUT_FN = "input_test.txt"
INPUT_FN = "input.txt"

with open(INPUT_FN) as f:
    lines = [l.rstrip() for l in f]
```

```python
# first line development

values = [int(s) for s in lines[0].split(' ')]
```

```python
values
```

```python
# Build up pyramid of differences down to zeros
diffs_pyr=[values]

while True:
    row = diffs_pyr[-1] # last row
    new_row = []
    is_all_zeros=True
    for i in range(len(row)-1):
        val=row[i+1]-row[i]
        new_row.append(val)
        if val!=0:
            is_all_zeros=False
    
    diffs_pyr.append(new_row)
    if is_all_zeros:
        break
```

Similar loop but different formula and elements to use

```python
new_vals = []
for sline in lines:
    values = [int(s) for s in sline.split(' ')]

    # Build up pyramid of differences down to zeros
    diffs_pyr=[values]
    while True:
        row = diffs_pyr[-1] # last row
        new_row = []
        is_all_zeros=True
        for i in range(len(row)-1):
            val=row[i+1]-row[i]
            new_row.append(val)
            if val!=0:
                is_all_zeros=False
        
        diffs_pyr.append(new_row)
        if is_all_zeros:
            break
    
    new_val_below = 0

    for i in range(len(diffs_pyr)-2, -1,-1): # start from previous to last row and go backwards
        row = diffs_pyr[i]
        #print(row)

        #first element and subtracts to value below
        new_val_above = row[0]- new_val_below

        new_val_below= new_val_above #next iteration

    #print(new_val_above)

    new_vals.append(new_val_above)
```
